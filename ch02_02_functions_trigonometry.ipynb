## 2.4 Functions

A **function** $f(x)$ is like a machine: you put in a number, and it gives you back exactly one output. The input is called the *argument* (usually $x$), and the output is the *value* of the function at that input.

> **Everyday analogy:** Think of a function like a vending machine. You press button A1 (input), and exactly one bag of crisps comes out (output). You can't get two different snacks from the same button, and you can't get nothing at all. That's what "exactly one output" means.

In quantum mechanics, **wavefunctions** are functions — typically of position $\psi(x)$ or time $\psi(t)$. The wavefunction contains all the information we can know about a quantum system.

Common functions you'll encounter include:
- **Polynomials** (like $x^2$, $x^3 + 2x - 1$)
- **Exponentials** (like $e^x$, $e^{-x^2}$)
- **Trigonometric functions** (like $\sin x$, $\cos x$)
- **Absolute value** (like $|x|$)
- **Logarithms** (like $\ln x$)

```python
# Polynomial function: f(x) = x² - 2x + 1
# This can be factored as f(x) = (x - 1)² — a parabola that touches the x-axis at x = 1

def f(x):
    return x**2 - 2*x + 1

print("f(x) = x² - 2x + 1:")
print("Notice how the outputs form a parabola shape:")
for x in [-2, -1, 0, 1, 2, 3]:
    print(f"  f({x:>2}) = {f(x):>2}")

# Key observations:
# - The minimum value is at x = 1, where f(1) = 0 (this is where the parabola "turns around")
# - The function is symmetric about x = 1: f(1 + d) = f(1 - d) for any distance d
#   For example: f(0) = f(2) = 1 and f(-1) = f(3) = 4

# Try factoring: x² - 2x + 1 = (x - 1)(x - 1) = (x - 1)²
# This means the only solution to f(x) = 0 is x = 1 (a "repeated root")
```

### Key properties of functions

Understanding these properties helps us classify and work with functions in quantum mechanics:

- **Even functions:** $f(-x) = f(x)$ — the graph is *symmetric* about the y-axis. Imagine folding the graph along the y-axis; both halves match perfectly. Common examples include $f(x) = x^2$ and $f(x) = \cos x$.

- **Odd functions:** $f(-x) = -f(x)$ — the graph is *symmetric* about the origin. If you rotate the graph by 180° around the origin, it looks the same. Common examples include $f(x) = x^3$ and $f(x) = \sin x$.

- **Periodic functions:** $f(x + T) = f(x)$ for some period $T$ — the graph repeats itself every $T$ units along the x-axis. The sine and cosine functions are periodic with period $2\pi \approx 6.283$. This repeating behaviour is crucial for describing waves.

> **Why care about even and odd functions?** In quantum mechanics, many wavefunctions are either even or odd. This symmetry tells us about the physical properties of the system, such as whether the particle is equally likely to be found on the left or right side of a potential well.

```python
# Test even/odd properties
def even_test(f, x):
    return f(-x) == f(x)

def odd_test(f, x):
    return f(-x) == -f(x)

x = 3
print(f"At x = {x}:")
print(f"  Even: f(x) = x² → f(-x) = {x**2}, f(-x) == f(x)? {even_test(lambda t: t**2, x)}")
print(f"  Odd:  f(x) = x³ → f(-x) = {(-x)**3}, f(-x) == -f(x)? {odd_test(lambda t: t**3, x)}")
```

## 2.5 Trigonometry

Trigonometric functions describe oscillations and waves — the heart of quantum mechanics. Every quantum particle behaves like a wave, and sine and cosine are the mathematical tools we use to describe that wave.

> **Why trigonometry?** You might remember SOH-CAH-TOA from school: sine, cosine, and tangent relate the angles of a right-angled triangle to the ratios of its sides. But this is just the starting point. In higher mathematics, we extend trigonometry to the **unit circle**, which lets us handle *any* angle — not just the acute angles in a triangle. This extension is what makes trigonometry powerful enough for quantum mechanics.

### The Unit Circle

On the unit circle (a circle with radius 1, centred at the origin), any point at angle $\theta$ measured counter-clockwise from the positive x-axis has coordinates:
$$(\cos\theta, \sin\theta)$$

This definition generalises what you learned in school. When $\theta = 30°$ (or $\pi/6$ radians), $\cos(30°) \approx 0.866$ and $\sin(30°) = 0.5$. When $\theta = 90°$ (or $\pi/2$ radians), we get $\cos(90°) = 0$ and $\sin(90°) = 1$.

> **Key concept: Radians vs degrees.** In mathematics and physics, angles are measured in **radians**, not degrees. One radian is the angle subtended when the arc length equals the radius. Since the full circumference of a unit circle is $2\pi$, a full rotation is $2\pi$ radians = 360°. Thus, $180° = \pi$ radians, $90° = \pi/2$ radians, and so on. You'll see radians everywhere in quantum mechanics.

```python
import math
# Angles in radians
angles = [0, math.pi/6, math.pi/4, math.pi/3, math.pi/2, math.pi]
angles_names = ["0", "30°", "45°", "60°", "90°", "180°"]

print(f"{'Angle':<10} {'Radians':<15} {'sin':<10} {'cos':<10}")
print("-" * 45)
for name, angle in zip(angles_names, angles):
    sin_val = round(math.sin(angle), 10)
    cos_val = round(math.cos(angle), 10)
    # Clean up representation
    sin_str = "0" if sin_val == 0 else f"{sin_val:.4f}"
    cos_str = "1" if angle == 0 else f"{cos_val:.4f}"
    print(f"  {name:<8}   {angle:.4f}     {sin_str:<10} {cos_str:<10}")
```

### The Pythagorean Identity

The most important trigonometric identity follows directly from the Pythagorean theorem applied to the unit circle:
$$\sin^2\theta + \cos^2\theta = 1$$

Here's why this is true: on the unit circle, any point $(\cos\theta, \sin\theta)$ is exactly distance 1 from the origin (that's what "unit circle" means!). By the Pythagorean theorem, the distance from the origin is $\sqrt{\cos^2\theta + \sin^2\theta} = 1$, so squaring both sides gives the identity.

> **Why does this matter in quantum mechanics?** This identity is the foundation for understanding normalisation — the requirement that the total probability of finding a particle *somewhere* must equal 1. The same mathematical structure appears: if you have two probabilities that add up to 1, they obey the same relationship as sine and cosine squared.

#### Other useful trigonometric identities

From this fundamental identity, we can derive others:
- $\tan\theta = \frac{\sin\theta}{\cos\theta}$
- $1 + \tan^2\theta = \sec^2\theta$
- $\sin(2\theta) = 2\sin\theta\cos\theta$
- $\cos(2\theta) = \cos^2\theta - \sin^2\theta$

These become useful when simplifying wavefunction equations.

```python
import math
# Verify the Pythagorean identity
theta = math.pi / 7  # Any angle works
sin_sq = math.sin(theta)**2
cos_sq = math.cos(theta)**2

print(f"θ = π/7 ≈ {theta:.4f}")
print(f"sin²(θ) = {sin_sq:.10f}")
print(f"cos²(θ) = {cos_sq:.10f}")
print(f"sin²(θ) + cos²(θ) = {sin_sq + cos_sq:.10f}")
print(f"Identity holds: {abs(sin_sq + cos_sq - 1) < 1e-10}")
```

### Euler's Formula — The Bridge to Quantum Mechanics

This is arguably the most important equation in all of quantum mechanics:
$$e^{i\theta} = \cos\theta + i\sin\theta$$

It connects three seemingly unrelated areas of mathematics — **exponentials**, **trigonometry**, and **complex numbers** — into one elegant relationship. To appreciate how remarkable this is, consider that each of these topics is usually taught in a completely separate maths class. Euler's formula shows they are deeply intertwined.

#### What does this formula *mean*?

Imagine a point rotating around the unit circle in the complex plane. The angle of rotation is $\theta$. Euler's formula tells us that this rotation can be described compactly as $e^{i\theta}$ — an exponential with an imaginary exponent. As $\theta$ changes, the point moves around the circle, tracing out $\cos\theta + i\sin\theta$.

> **A visual way to think about it:** 
> - The real part of $e^{i\theta}$ (the horizontal coordinate) is $\cos\theta$
> - The imaginary part of $e^{i\theta}$ (the vertical coordinate) is $\sin\theta$
> - The distance from the origin is always 1 (it's on the unit circle)

#### Euler's Identity: The Special Case

When $\theta = \pi$, we get:
$$e^{i\pi} = \cos\pi + i\sin\pi = -1 + i(0) = -1$$

Rearranging gives the famous **Euler's identity**:
$$e^{i\pi} + 1 = 0$$

This equation is remarkable because it links together five of the most important numbers in mathematics:
- $e$ (from calculus and exponential growth)
- $i$ (from complex numbers)
- $\pi$ (from geometry and circles)
- $1$ (the multiplicative identity)
- $0$ (the additive identity)

In one elegant equation!

```python
import cmath
import math
# Euler's formula
theta = math.pi / 3

# Left side: e^(iθ)
euler_lhs = cmath.exp(1j * theta)

# Right side: cos(θ) + i·sin(θ)
euler_rhs = math.cos(theta) + 1j * math.sin(theta)

print(f"θ = π/3 = {theta:.4f}")
print(f"e^(iθ) = {euler_lhs}")
print(f"cos(θ) + i·sin(θ) = {euler_rhs}")
print(f"They match: {abs(euler_lhs - euler_rhs) < 1e-14}")
```

```python
import cmath
import math
# Verify Euler's identity: e^(iπ) + 1 = 0
eulers_identity = cmath.exp(1j * math.pi) + 1
print(f"e^(iπ) + 1 = {eulers_identity}")
print(f"Close to zero: {abs(eulers_identity) < 1e-14}")
```
