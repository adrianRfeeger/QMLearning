## 2.8 Vectors and Matrices (Linear Algebra)

By this point, we've covered arithmetic, algebra, functions, trigonometry, complex numbers, and probability. Now comes the final piece: **linear algebra** — the study of vectors and matrices. Linear algebra is the primary mathematical language of quantum mechanics, and it might feel very different from the algebra you learned in school.

> **A shift in perspective:** In school algebra, you solve for *numbers* (like $x = 5$). In linear algebra, you solve for *vectors* and *matrices*. Instead of asking "what number satisfies this equation?" you ask "what vector satisfies this equation?" This shift in thinking is essential for quantum mechanics.

A quantum state is represented as a **vector** in a complex vector space (called a **Hilbert space**). Physical operations (like measuring energy or applying a magnetic field) are represented by **matrices** that transform these vectors.

### Column Vectors and Ket Notation

Physicist Paul Dirac invented a brilliant notation called **bra-ket notation** (pronounced "bra" and "ket") to keep track of vectors and their partners:
- A **ket** $|\psi\rangle$ represents a column vector (think of it as the right half of a bracket $\rangle$).
- A **bra** $\langle\psi|$ represents the conjugate transpose — a row vector containing the complex conjugates (think of it as the left half of a bracket $\langle$).

Together, $\langle\phi|\psi\rangle$ forms a "bracket" representing the inner product (like a dot product) between two vectors.

The two fundamental qubit states — the quantum analogue of 0 and 1 in classical computing — are written as:

```python
import numpy as np

# Define the computational basis states
ket_0 = np.array([[1],
                  [0]])
ket_1 = np.array([[0],
                  [1]])

print("State |0> = ", ket_0.T)
print("State |1> = ", ket_1.T)
print("\nWritten as column vectors:")
print("|0⟩ = [1, 0]^T")
print("|1⟩ = [0, 1]^T")
```

### Inner Products (Bra·Ket)

The inner product $\langle\phi|\psi\rangle$ tells us how "aligned" two states are — it's the quantum version of asking "how similar are these two vectors?"

> **Analogy:** Imagine two arrows on a piece of paper. If they point in exactly the same direction, they are perfectly aligned. If they point in perpendicular directions, they are completely unrelated. The inner product quantifies this idea mathematically.

Key properties:
- **Orthogonal states:** If $\langle\phi|\psi\rangle = 0$, the states are *orthogonal* (perpendicular). In quantum mechanics, this means the states are completely distinguishable. If the system is in state $|\psi\rangle$, a measurement will *never* find it in state $|\phi\rangle$.
- **Normalised states:** If $\langle\psi|\psi\rangle = 1$, the state is normalised — the total probability of finding the particle *somewhere* is 100%.
- **Orthonormal sets:** If $\langle i|j\rangle = \delta_{ij}$ (the Kronecker delta, equal to 1 when $i=j$ and 0 otherwise), the states form an orthonormal basis — the quantum analogue of the x and y axes in 2D space.

For the qubit states $|0\rangle$ and $|1\rangle$:
- $\langle 0|0\rangle = 1$ (|0⟩ is normalised to itself)
- $\langle 1|1\rangle = 1$ (|1⟩ is normalised to itself)
- $\langle 0|1\rangle = 0$ (|0⟩ and |1⟩ are orthogonal — they are perfectly distinguishable)

This orthonormality is fundamental: it guarantees that when we measure a qubit, we get exactly one definite outcome (either 0 or 1), never both and never neither.

```python
# The bra is the conjugate transpose of the ket
# In numpy: H is conjugate transpose (Hermitian transpose)
bra_0 = ket_0.T.conj()  # <0|
bra_1 = ket_1.T.conj()  # <1|

# Inner products
overlap_01 = bra_0 @ ket_1  # <0|1>
overlap_00 = bra_0 @ ket_0  # <0|0>
overlap_11 = bra_1 @ ket_1  # <1|1>

print(f"<0|0> = {overlap_00} (orthonormal: yes)")
print(f"<1|1> = {overlap_11} (orthonormal: yes)")
print(f"<0|1> = {overlap_01} (orthogonal: yes)")
```

### Vector Norm (Normalization)

The **norm** (pronounced "norm" as in "standard") of a vector measures its length — how far it stretches from the origin. For a quantum state, the norm must equal 1. This is not just a mathematical convenience; it has a deep physical meaning:

> **The normalisation condition:** $\langle\psi|\psi\rangle = 1$ means the total probability of finding the particle *somewhere* is 100%. If the norm were less than 1, there would be less than 100% chance of finding the particle. If it were greater than 1, the probabilities would be nonsense.

For a vector $|\psi\rangle = \begin{pmatrix} a \\ b \end{pmatrix}$, the norm squared is:
$$\langle\psi|\psi\rangle = |a|^2 + |b|^2$$

This is simply the sum of the probabilities of all possible outcomes — and it must equal 1.

#### How to normalise a vector

If you have a vector that isn't normalised, you can fix it by dividing by its norm. For example, the vector $\begin{pmatrix} 3 \\ 4 \end{pmatrix}$ has norm $\sqrt{3^2 + 4^2} = \sqrt{25} = 5$. Dividing by 5 gives the normalised vector $\begin{pmatrix} 3/5 \\ 4/5 \end{pmatrix} = \begin{pmatrix} 0.6 \\ 0.8 \end{pmatrix}$.

> **Connection to the Pythagorean theorem:** The normalisation condition $|a|^2 + |b|^2 = 1$ is the same structure as the Pythagorean identity $\sin^2\theta + \cos^2\theta = 1$. If we write $a = \cos\theta$ and $b = \sin\theta$, the normalisation condition is automatically satisfied. This is no coincidence — both express the idea that "the parts add up to the whole".

```python
# Check normalization of |0⟩ and |1⟩
norm_0 = np.sqrt(np.real((bra_0 @ ket_0).real))
norm_1 = np.sqrt(np.real((bra_1 @ ket_1).real))
print(f"||0⟩|| = {norm_0:.4f}")
print(f"||1⟩|| = {norm_1:.4f}")
print(f"Both normalized: {abs(norm_0 - 1) < 1e-10 and abs(norm_1 - 1) < 1e-10}")
```

### Superposition

**Superposition** is perhaps the most famous (and most misunderstood) concept in quantum mechanics. It means that a quantum system can exist in a combination of multiple states simultaneously — not as one *or* the other, but as *both at the same time*.

A quantum system in superposition is written as:
$$|\psi\rangle = \alpha |0\rangle + \beta |1\rangle$$

where $\alpha$ and $\beta$ are complex numbers called **probability amplitudes**, and they must satisfy the normalisation condition:
$$|\alpha|^2 + |\beta|^2 = 1$$

> **The cat analogy (and why it's misleading):** Erwin Schrödinger proposed a thought experiment involving a cat in a sealed box with a radioactive atom. If the atom decays, the cat dies; if it doesn't, the cat lives. According to quantum mechanics, until we open the box, the cat is in a *superposition* of alive and dead. While this is a popular illustration, it's actually a *reductio ad absurdum* — Schrödinger meant it to show how absurd it sounds, not to claim cats can actually be both alive and dead. Quantum superpositions are very fragile and don't scale up to everyday objects like cats.

#### What does "both at once" really mean?

When we say the qubit is in a superposition of $|0\rangle$ and $|1\rangle$, we mean:
1. The state *exists* as a definite mathematical object: $|\psi\rangle = \alpha |0\rangle + \beta |1\rangle$.
2. When we *measure* the qubit, we only ever see one outcome: either 0 (with probability $|\alpha|^2$) or 1 (with probability $|\beta|^2$).
3. The superposition is revealed when we perform *multiple* measurements on identically prepared systems — the statistics match the probabilities $|\alpha|^2$ and $|\beta|^2$.

> **Key insight:** The superposition is real, but the act of measurement *collapses* it to one definite outcome. Before measurement, the qubit is genuinely in both states. After measurement, it settles into whichever outcome was observed. This collapse is one of the deepest mysteries of quantum mechanics.

#### The role of phase in superposition

Notice that $\alpha$ and $\beta$ are *complex* numbers — they have both a magnitude and a phase. The phase doesn't affect the probabilities of individual outcomes (since $|\alpha|^2$ only depends on the magnitude), but it *does* affect how different parts of the superposition *interfere* with each other when operations are applied. This interference is what gives quantum computers their power.

```python
# Create a superposition state
alpha = 1 / np.sqrt(2)
beta = 1j / np.sqrt(2)  # beta has a phase (imaginary part)

# Verify normalization: |α|² + |β|² = 1
norm_sq = abs(alpha)**2 + abs(beta)**2
print(f"Superposition: |ψ⟩ = {alpha}|0⟩ + {beta}i|1⟩")
print(f"Normality check: |α|² + |β|² = {norm_sq:.4f}")

# The ket vector
ket_psi = alpha * ket_0 + beta * ket_1
print(f"\n|ψ> = {ket_psi.flatten()}")

# Probabilities of each outcome
p0 = abs(alpha)**2
p1 = abs(beta)**2
print(f"\nIf we measure:")
print(f"  P(0) = |α|² = {p0:.4f} → {p0 * 100:.0f}%")
print(f"  P(1) = |β|² = {p1:.4f} → {p1 * 100:.0f}%")
print(f"  Total = {p0 + p1:.4f}")
```

### Quantum Operators (Matrices)

In quantum mechanics, **operations** — such as measuring energy, applying a magnetic field, or flipping a qubit — are represented by **matrices**. These matrices transform one quantum state into another, just like a function transforms one number into another.

> **A shift in perspective:** In school, you probably encountered functions like $f(x) = x^2$. In quantum mechanics, the "functions" are *linear operators* — matrices that act on vectors. When we apply an operator to a state vector, we get a new state vector. This is the mathematical language of "doing something" to a quantum system.

#### What does a matrix actually *do*?

A matrix is just a rectangular array of numbers. But when you multiply a matrix by a vector, it transforms the vector — it can stretch it, rotate it, reflect it, or squish it. Here are some intuitive examples:

- **Scaling:** The identity matrix $\begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$ does nothing — it leaves every vector unchanged. It's like multiplying by 1.
- **Swapping:** The Pauli-X matrix $\begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}$ flips the vector — it swaps the top and bottom components. If the input is $|0\rangle$, the output is $|1\rangle$, and vice versa. This is the quantum equivalent of a "NOT" gate.
- **Rotation:** Some matrices rotate vectors in the complex plane. These are crucial for creating and manipulating superpositions.

#### Eigenvalues and eigenvectors

When a matrix operator $M$ acts on a special kind of vector (called an **eigenvector**), it doesn't change the direction — it only stretches or shrinks the vector by a factor called the **eigenvalue**:
$$M|\lambda\rangle = \lambda |\lambda\rangle$$

> **Why this matters in quantum mechanics:** The eigenvectors of an operator represent the states in which a measurement gives a *definite* result (no superposition). The eigenvalues are the possible measurement outcomes. For example, the energy levels of an atom are the eigenvalues of the Hamiltonian (energy) operator.

#### Physical observables as Hermitian operators

In quantum mechanics, every measurable quantity (called an **observable**) — energy, position, momentum, spin — is represented by a **Hermitian matrix**. Hermitian matrices have two crucial properties:
1. All eigenvalues are **real numbers** — measurement outcomes must be real, not complex.
2. Eigenvectors corresponding to different eigenvalues are **orthogonal** — different outcomes are perfectly distinguishable.

The Pauli matrices (X, Y, Z) are the most important Hermitian operators in quantum computing, and they represent the three spatial directions in which we can measure a qubit's spin.

```python
# Pauli-X matrix
X_gate = np.array([[0, 1],
                   [1, 0]])

# Apply X to |0>
X_on_0 = X_gate @ ket_0
print("X|0⟩ =", X_on_0.flatten())
print("Expected: |1⟩ = [0, 1]")
print(f"Match: {np.array_equal(np.round(X_on_0.flatten(), 10), ket_1.flatten())}")

# Apply X to |1>
X_on_1 = X_gate @ ket_1
print(f"\nX|1⟩ = {X_on_1.flatten()}")
print("Expected: |0⟩ = [1, 0]")
print(f"Match: {np.array_equal(np.round(X_on_1.flatten(), 10), ket_0.flatten())}")
```

### The Hadamard Gate

The **Hadamard gate** is perhaps the most important quantum gate because it is the standard way to *create* a superposition from a definite state. It is denoted by the letter $H$ and is defined as:
$$H = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 & 1 \\ 1 & -1 \end{pmatrix}$$

> **Why the factor of $\frac{1}{\sqrt{2}}$?** This normalisation factor ensures that the output states are properly normalised (have unit length). Without it, the probabilities wouldn't add up to 1, which would violate the fundamental rules of quantum mechanics.

When applied to $|0\rangle$, the Hadamard gate creates an *equal* superposition:
$$H|0\rangle = \frac{|0\rangle + |1\rangle}{\sqrt{2}}$$

This means if we measure the qubit, we have a 50% chance of getting 0 and a 50% chance of getting 1. It's the quantum analogue of flipping a fair coin.

When applied to $|1\rangle$, it creates a *different* equal superposition:
$$H|1\rangle = \frac{|0\rangle - |1\rangle}{\sqrt{2}}$$

Notice the **minus sign** in front of $|1\rangle$. This is not just a cosmetic difference — the minus sign represents a *phase difference*. In quantum mechanics, phase matters enormously. Two superpositions that look similar can have very different behaviours because of subtle phase differences. This is the source of quantum interference, which is exploited in quantum algorithms to amplify correct answers and cancel wrong ones.

#### The Hadamard transform is its own inverse

Applying the Hadamard gate twice brings you back to where you started:
$$H \cdot H = I$$
where $I$ is the identity matrix. This is a special property — not all quantum gates are reversible in this way. It also means that if you apply $H$ to a superposition state, you can "undo" the superposition and recover the original definite state.

> **Why is this important?** It means quantum operations are *reversible* — you can always go backwards. This is fundamentally different from classical computing, where operations like AND and OR are irreversible (you can't tell what the inputs were just by looking at the output). Reversibility is a key feature of quantum mechanics and has deep implications for quantum computing and thermodynamics.

```python
# Hadamard gate
H = (1 / np.sqrt(2)) * np.array([[1, 1],
                                 [1, -1]])

H_on_0 = H @ ket_0
H_on_1 = H @ ket_1

print("H|0⟩ =", H_on_0.flatten())
print("H|1⟩ =", H_on_1.flatten())

# Verify both are normalized
norm_H0 = np.sqrt(np.real(H_on_0.T.conj() @ H_on_0))
norm_H1 = np.sqrt(np.real(H_on_1.T.conj() @ H_on_1))
print(f"\n||H|0⟩|| = {norm_H0:.4f}")
print(f"||H|1⟩|| = {norm_H1:.4f}")
```

### The big picture

Quantum mechanics can be understood through a few central ideas, all built on the mathematical foundations we've developed:

1. **States are vectors.** A quantum system is described by a vector $|\psi\rangle$ in a complex vector space.
2. **Operations are matrices.** Physical processes transform states via matrix multiplication.
3. **Probabilities come from amplitudes.** The probability of an outcome is the squared magnitude of a complex amplitude: $P = |\psi|^2$.
4. **Phase matters.** Complex numbers carry phase information, which leads to interference — the hallmark of quantum behaviour.

These ideas may seem abstract now, but they are straightforward consequences of the mathematics we've explored. The next chapters will show you how these tools are applied to understand real quantum phenomena.
