# 4. Spin and Qubits

We have discussed a particle moving in space, which has an infinite number of possible positions. But some quantum systems are much simpler. They only have **two** possible states.

## 4.1 Quantum Spin

In 1922, the Stern-Gerlach experiment fired a beam of silver atoms through a nonuniform magnetic field. Instead of the beam spreading out continuously (as classical physics predicted), it split into two distinct paths: "Spin Up" and "Spin Down". 

This was the discovery of an intrinsic quantum property called **spin**. It's somewhat analogous to a planet spinning on its axis, but it's fundamentally a purely quantum phenomenon. For an electron, measuring its spin along any axis will always yield one of two results: $+\hbar/2$ (up) or $-\hbar/2$ (down).

## 4.2 The Qubit

In classical computing, a bit is either a `0` or a `1`. In quantum computing, we use a two-state quantum system as our basic unit of information, called a **qubit**.

We denote the two basic states as $|0\rangle$ (analogous to classical 0 or spin up) and $|1\rangle$ (analogous to classical 1 or spin down). Unlike a classical bit, a qubit can be in a **superposition** of these states:

$$|\psi\rangle = \alpha |0\rangle + \beta |1\rangle$$

where $\alpha$ and $\beta$ are complex numbers such that their probabilities sum to 1: $|\alpha|^2 + |\beta|^2 = 1$.

## 4.3 The Bloch Sphere

Because the state of a qubit depends on complex numbers, it can be visualized as a point on the surface of a 3D sphere called the **Bloch Sphere**. 
- The North Pole represents the state $|0\rangle$.
- The South Pole represents the state $|1\rangle$.
- Points on the equator represent equal superpositions of $|0\rangle$ and $|1\rangle$ (like the $|+\rangle$ state).

Let's visualize some qubit states using Python's `matplotlib`.

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_bloch_vector(ax, theta, phi, label, color='b'):
    """
    Plots a vector on the Bloch sphere given spherical coordinates.
    theta: angle from the Z axis (0 to pi)
    phi: angle from the X axis in the XY plane (0 to 2pi)
    """
    # Convert spherical coordinates to Cartesian coordinates for plotting
    x = np.sin(theta) * np.cos(phi)
    y = np.sin(theta) * np.sin(phi)
    z = np.cos(theta)
    
    # Draw the vector
    ax.quiver(0, 0, 0, x, y, z, color=color, arrow_length_ratio=0.1, linewidth=2)
    # Add the label slightly further out from the vector tip
    ax.text(x * 1.2, y * 1.2, z * 1.2, label, color=color, fontsize=14, fontweight='bold')

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')

# Draw the sphere surface
u = np.linspace(0, 2 * np.pi, 50)
v = np.linspace(0, np.pi, 50)
X = np.outer(np.cos(u), np.sin(v))
Y = np.outer(np.sin(u), np.sin(v))
Z = np.outer(np.ones(np.size(u)), np.cos(v))

# Plot a wireframe sphere
ax.plot_wireframe(X, Y, Z, color="gray", alpha=0.1)
ax.plot_surface(X, Y, Z, color="cyan", alpha=0.05)

# Plot the X, Y, Z axes inside the sphere
ax.plot([-1.1, 1.1], [0, 0], [0, 0], color='k', linestyle='--', linewidth=1)
ax.plot([0, 0], [-1.1, 1.1], [0, 0], color='k', linestyle='--', linewidth=1)
ax.plot([0, 0], [0, 0], [-1.1, 1.1], color='k', linestyle='--', linewidth=1)

# Label the axes
ax.text(1.2, 0, 0, 'X', color='k', fontsize=12)
ax.text(0, 1.2, 0, 'Y', color='k', fontsize=12)
ax.text(0, 0, 1.2, 'Z', color='k', fontsize=12)

# Plot some important states
# State |0> (North Pole: theta=0, phi=0)
plot_bloch_vector(ax, 0, 0, '|0>', color='red')

# State |1> (South Pole: theta=pi, phi=0)
plot_bloch_vector(ax, np.pi, 0, '|1>', color='blue')

# State |+> (Equator on X axis: theta=pi/2, phi=0)
# This is the equal superposition state (1/sqrt(2))(|0> + |1>)
plot_bloch_vector(ax, np.pi/2, 0, '|+>', color='green')

# State |i> (Equator on Y axis: theta=pi/2, phi=pi/2)
# This is the equal superposition state with an imaginary phase (1/sqrt(2))(|0> + i|1>)
plot_bloch_vector(ax, np.pi/2, np.pi/2, '|i>', color='purple')

# Formatting
ax.set_xlim([-1, 1])
ax.set_ylim([-1, 1])
ax.set_zlim([-1, 1])
ax.set_box_aspect([1,1,1])
ax.set_axis_off() # Hide the background grid for a cleaner look
plt.title('The Bloch Sphere representation of a Qubit', fontsize=16)
plt.tight_layout()
plt.show()
```

When we perform operations on a qubit in a quantum computer (like applying the Pauli-X gate we saw in Chapter 2), it corresponds to **rotating** this vector around the axes of the Bloch Sphere.

- The **Pauli-X gate** is a 180-degree rotation around the X-axis. Notice how that would flip the red vector $|0\rangle$ down to the blue vector $|1\rangle$!
- The **Pauli-Z gate** is a 180-degree rotation around the Z-axis. It leaves $|0\rangle$ and $|1\rangle$ unchanged, but would flip the green vector $|+\rangle$ to the opposite side of the sphere.
