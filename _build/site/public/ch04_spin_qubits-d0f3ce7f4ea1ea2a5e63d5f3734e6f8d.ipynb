# 4. Spin and Qubits

We have discussed a particle moving in space, which has an infinite number of possible positions. But some quantum systems are much simpler. They only have **two** possible states.

## 4.1 Quantum Spin

In 1922, the Stern-Gerlach experiment fired a beam of silver atoms through a nonuniform magnetic field. Instead of the beam spreading out continuously (as classical physics predicted), it split into two distinct paths: "Spin Up" and "Spin Down". 

This was the discovery of an intrinsic quantum property called **spin**. For an electron, measuring its spin along any axis will always yield one of two results: $+\hbar/2$ (up) or $-\hbar/2$ (down).

## 4.2 The Qubit

In classical computing, a bit is either a `0` or a `1`. In quantum computing, we use a two-state quantum system as our basic unit of information, called a **qubit**.

We denote the two basic states as $|0\rangle$ and $|1\rangle$. Unlike a classical bit, a qubit can be in a **superposition** of these states:

$$|\psi\rangle = \alpha |0\rangle + \beta |1\rangle$$

where $\alpha$ and $\beta$ are complex numbers such that $|\alpha|^2 + |\beta|^2 = 1$.

## 4.3 The Bloch Sphere

Because the state of a qubit depends on complex numbers, it can be visualized as a point on the surface of a 3D sphere called the **Bloch Sphere**. 
- The North Pole represents the state $|0\rangle$.
- The South Pole represents the state $|1\rangle$.
- Points on the equator represent equal superpositions of $|0\rangle$ and $|1\rangle$.

Let's visualize some qubit states! (We will just plot points on a 3D sphere using matplotlib to simulate this conceptually).

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_bloch_vector(theta, phi, label, color='b'):
    # Convert spherical coordinates to Cartesian coordinates for plotting
    x = np.sin(theta) * np.cos(phi)
    y = np.sin(theta) * np.np.sin(phi)
    z = np.cos(theta)
    
    ax.quiver(0, 0, 0, x, y, z, color=color, arrow_length_ratio=0.1)
    ax.text(x * 1.15, y * 1.15, z * 1.15, label, color=color, fontsize=12, fontweight='bold')

fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Draw the sphere
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)
X = np.outer(np.cos(u), np.sin(v))
Y = np.outer(np.sin(u), np.sin(v))
Z = np.outer(np.ones(np.size(u)), np.cos(v))
ax.plot_wireframe(X, Y, Z, color="k", alpha=0.1)
ax.plot_surface(X, Y, Z, color="c", alpha=0.05)

# Plot the axes
ax.plot([-1, 1], [0, 0], [0, 0], color='k', linestyle='--', linewidth=0.5)
ax.plot([0, 0], [-1, 1], [0, 0], color='k', linestyle='--', linewidth=0.5)
ax.plot([0, 0], [0, 0], [-1, 1], color='k', linestyle='--', linewidth=0.5)
ax.text(1.2, 0, 0, 'X', color='k')
ax.text(0, 1.2, 0, 'Y', color='k')
ax.text(0, 0, 1.2, 'Z', color='k')

# State |0> (North Pole)
plot_bloch_vector(0, 0, '|0>', color='r')

# State |1> (South Pole)
plot_bloch_vector(np.pi, 0, '|1>', color='b')

# State |+> (Equator on X axis)
plot_bloch_vector(np.pi/2, 0, '|+>', color='g')

ax.set_box_aspect([1,1,1])
ax.set_axis_off()
plt.title('The Bloch Sphere')
plt.show()
```

When we perform operations on a qubit in a quantum computer (like applying the Pauli-X gate we saw in Chapter 2), it corresponds to **rotating** this vector around the axes of the Bloch Sphere.
