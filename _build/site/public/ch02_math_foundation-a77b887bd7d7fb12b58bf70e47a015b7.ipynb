# 2. Mathematical Foundations

To truly understand quantum mechanics, we must speak its language. Unlike classical physics, which relies heavily on calculus to track the continuous motion of objects, quantum mechanics relies primarily on **linear algebra** and **complex numbers**.

This chapter builds the required mathematical tools from the ground up — starting with arithmetic and algebra, progressing through trigonometry and functions, and culminating in the vector spaces and complex numbers that form the backbone of quantum theory.

## 2.1 Arithmetic and Number Systems

### Integers, Fractions, and Decimals

We begin with the most basic number systems. **Integers** ($\mathbb{Z} = \dots, -2, -1, 0, 1, 2, \dots$) are the counting numbers extended to include negatives. **Rational numbers** ($\mathbb{Q}$) are any numbers expressible as a fraction $\frac{p}{q}$ where $p$ and $q$ are integers and $q \neq 0$.

```python
# Basic arithmetic operations
a, b = 7, 3

print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} × {b} = {a * b}")
print(f"Division: {a} ÷ {b} = {a / b}")
print(f"Integer division: {a} // {b} = {a // b}")
print(f"Remainder: {a} % {b} = {a % b}")
print(f"Exponentiation: {a} ^ {b} = {a ** b}")
```

### Irrational Numbers and $\pi$

Not all numbers can be written as fractions. **Irrational numbers** like $\pi$ and $\sqrt{2}$ have non-repeating, non-terminating decimal expansions. They are essential for describing circles, waves, and oscillations — all central to quantum mechanics.

```python
import math

# The number pi
print(f"π ≈ {math.pi:.15f}")

# Square roots of non-perfect squares are irrational
print(f"√2 ≈ {math.sqrt(2):.15f}")
print(f"√3 ≈ {math.sqrt(3):.15f}")

# The imaginary unit: sqrt(-1), which we'll revisit shortly
print(f"√(-1) is the imaginary unit i (or j in Python)")
```

## 2.2 Exponents and Logarithms

Exponentials and logarithms appear everywhere in quantum mechanics — from wave amplitudes to probability decays.

**Exponent rules:**
$$a^m \cdot a^n = a^{m+n}, \quad \frac{a^m}{a^n} = a^{m-n}, \quad (a^m)^n = a^{m \cdot n}$$

The **natural exponential function** $e^x$ (where $e \approx 2.718$) is the most important function in quantum mechanics because it naturally arises from differential equations.

```python
# Exponentials
base, exp1, exp2 = 2, 3, 5

print(f"{base}^{exp1} = {base ** exp1}")
print(f"{base}^{exp2} = {base ** exp2}")

# Verify exponent rule: a^m * a^n = a^(m+n)
print(f"\n{base}^{exp1} × {base}^{exp2} = {base ** exp1 * base ** exp2}")
print(f"{base}^({exp1}+{exp2}) = {base ** (exp1 + exp2)}")

# The natural exponential
print(f"\ne ≈ {math.e:.15f}")
print(f"e^1 = {math.exp(1)}")
print(f"e^0 = {math.exp(0)}")
print(f"e^2 ≈ {math.exp(2):.4f}")
print(f"e^(-1) ≈ {math.exp(-1):.4f}")
```

### Logarithms

The natural logarithm $\ln(x)$ is the inverse of $e^x$, meaning $\ln(e^x) = x$.

```python
x = math.exp(3)
log_x = math.log(x)  # natural log

print(f"e^3 ≈ {x:.6f}")
print(f"ln(e^3) = {log_x}")
print(f"So ln and e^ cancel each other out: {abs(log_x - 3) < 1e-10}")

# Log rules: ln(ab) = ln(a) + ln(b)
a, b = 5, 7
print(f"\nln({a} × {b}) = ln({a * b}) = {math.log(a * b):.6f}")
print(f"ln({a}) + ln({b}) = {math.log(a) + math.log(b):.6f}")
```

## 2.3 Algebra and Equations

Algebra lets us solve for unknown quantities. In quantum mechanics, we frequently solve equations for eigenvalues and eigenvectors. But first, let's review the basics.

### Solving Linear Equations

Given $ax + b = 0$, we solve for $x = -\frac{b}{a}$.

```python
# Solve ax + b = c for x: x = (c - b) / a
a, b, c = 3, 7, 16

x = (c - b) / a
print(f"{a}x + {b} = {c} → x = (c - b) / a = ({c} - {b}) / {a} = {x}")

# Verify
print(f"Check: {a}({x}) + {b} = {a * x + b}")
```

### Quadratic Equations

Given $ax^2 + bx + c = 0$, the quadratic formula gives:
$$x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$$

```python
# Solve ax^2 + bx + c = 0
a, b, c = 1, -5, 6  # x^2 - 5x + 6 = 0, roots at x=2 and x=3

discriminant = b**2 - 4*a*c
root1 = (-b + math.sqrt(discriminant)) / (2*a)
root2 = (-b - math.sqrt(discriminant)) / (2*a)

print(f"{a}x² + ({b})x + {c} = 0")
print(f"Discriminant: Δ = b² - 4ac = {discriminant}")
print(f"Root 1: x = {root1}")
print(f"Root 2: x = {root2}")

# Verify
print(f"\nCheck root 1: {a}*({root1})² + {b}*({root1}) + {c} = {a*root1**2 + b*root1 + c}")
print(f"Check root 2: {a}*({root2})² + {b}*({root2}) + {c} = {a*root2**2 + b*root2 + c}")
```

### What if the discriminant is negative?

When $b^2 - 4ac < 0$, the square root involves **negative numbers**, which leads us to **complex numbers** — the language of quantum amplitudes.

```python
# Solve x^2 + 4 = 0, which has no real solution
a2, b2, c2 = 1, 0, 4  # x^2 + 4 = 0

discriminant2 = b2**2 - 4*a2*c2
print(f"{a2}x² + {b2}x + {c2} = 0")
print(f"Discriminant: Δ = {discriminant2} (negative → complex roots)")

# In Python, we use complex numbers
root1 = complex(0, math.sqrt(abs(discriminant2))) / (2*a2) * -b2
root2 = -root1
print(f"Roots: x = 0 ± 2i")
```

## 2.4 Functions

A **function** $f(x)$ assigns exactly one output to each input. In quantum mechanics, wavefunctions are functions — typically of position $\psi(x)$ or time $\psi(t)$.

Common functions include polynomials, exponentials, and trigonometric functions.

```python
# Polynomial function: f(x) = x² - 2x + 1
def f(x):
    return x**2 - 2*x + 1

print("f(x) = x² - 2x + 1:")
for x in [-2, -1, 0, 1, 2, 3]:
    print(f"  f({x}) = {f(x)}")
```

### Key properties of functions

- **Even functions:** $f(-x) = f(x)$ (symmetric about the y-axis), e.g., $x^2$
- **Odd functions:** $f(-x) = -f(x)$ (symmetric about the origin), e.g., $x^3$
- **Periodic functions:** $f(x + T) = f(x)$ for some period $T$, e.g., $\sin(x)$

```python
# Test even/odd properties
def even_test(f, x):
    return f(-x) == f(x)

def odd_test(f, x):
    return f(-x) == -f(x)

x = 3
print(f"At x = {x}:")
print(f"  Even: f(x) = x² → f(-x) = {x**2}, f(-x) == f(x)? {even_test(lambda t: t**2, x)}")
print(f"  Odd:  f(x) = x³ → f(-x) = {(-x)**3}, f(-x) == -f(x)? {odd_test(lambda t: t**3, x)}")
```

## 2.5 Trigonometry

Trigonometric functions describe oscillations and waves — the heart of quantum mechanics. The wavefunction is fundamentally a wave-like mathematical object.

### The Unit Circle

On the unit circle (radius 1, centered at origin), any point at angle $\theta$ has coordinates:
$$(\cos\theta, \sin\theta)$$

```python
# Angles in radians
angles = [0, math.pi/6, math.pi/4, math.pi/3, math.pi/2, math.pi]
angles_names = ["0", "30°", "45°", "60°", "90°", "180°"]

print(f"{'Angle':<10} {'Radians':<15} {'sin':<10} {'cos':<10}")
print("-" * 45)
for name, angle in zip(angles_names, angles):
    sin_val = round(math.sin(angle), 10)
    cos_val = round(math.cos(angle), 10)
    # Clean up representation
    sin_str = "0" if sin_val == 0 else f"{sin_val:.4f}"
    cos_str = "1" if angle == 0 else f"{cos_val:.4f}"
    print(f"  {name:<8}   {angle:.4f}     {sin_str:<10} {cos_str:<10}")
```

### The Pythagorean Identity

The most important trigonometric identity follows from the Pythagorean theorem applied to the unit circle:
$$\sin^2\theta + \cos^2\theta = 1$$

```python
# Verify the Pythagorean identity
theta = math.pi / 7  # Any angle works
sin_sq = math.sin(theta)**2
cos_sq = math.cos(theta)**2

print(f"θ = π/7 ≈ {theta:.4f}")
print(f"sin²(θ) = {sin_sq:.10f}")
print(f"cos²(θ) = {cos_sq:.10f}")
print(f"sin²(θ) + cos²(θ) = {sin_sq + cos_sq:.10f}")
print(f"Identity holds: {abs(sin_sq + cos_sq - 1) < 1e-10}")
```

### Euler's Formula — The Bridge to Quantum Mechanics

This is arguably the most important equation in all of quantum mechanics:
$$e^{i\theta} = \cos\theta + i\sin\theta$$

It connects exponentials, trigonometry, and complex numbers into one elegant relationship. Euler's identity $e^{i\pi} + 1 = 0$ is the special case where $\theta = \pi$.

```python
# Euler's formula
theta = math.pi / 3

# Left side: e^(iθ)
euler_lhs = cmath.exp(1j * theta)

# Right side: cos(θ) + i·sin(θ)
euler_rhs = math.cos(theta) + 1j * math.sin(theta)

print(f"θ = π/3 = {theta:.4f}")
print(f"e^(iθ) = {euler_lhs}")
print(f"cos(θ) + i·sin(θ) = {euler_rhs}")
print(f"They match: {abs(euler_lhs - euler_rhs) < 1e-14}")
```

```python
# Verify Euler's identity: e^(iπ) + 1 = 0
eulers_identity = cmath.exp(1j * math.pi) + 1
print(f"e^(iπ) + 1 = {eulers_identity}")
print(f"Close to zero: {abs(eulers_identity) < 1e-14}")
```

### Sine waves as quantum building blocks

A traveling wave can be written as $\psi(x,t) = A \sin(kx - \omega t + \phi)$, where:
- $A$ is the amplitude
- $k$ is the wave number (related to wavelength: $k = 2\pi/\lambda$)
- $\omega$ is the angular frequency ($\omega = 2\pi f$)
- $\phi$ is the phase

```python
import numpy as np

# Show how sine waves look
x = np.linspace(0, 2 * math.pi, 200)
wave = np.sin(x)

# Show a few values
sample_pts = [0, 30, 60, 100, 130, 169, 199]
print(f"{'x/π':>8} {'sin(x)':>10}")
print("-" * 20)
for idx in sample_pts:
    xVal = x[idx]
    print(f"{xVal/math.pi:6.2f}π  {wave[idx]:10.4f}")
```

## 2.6 Probability and Statistics Basics

Quantum mechanics is inherently probabilistic. We quantify outcomes using probability distributions and statistical measures.

### Discrete Probability

For a discrete set of outcomes, if outcome $i$ has probability $p_i$, then:
- $0 \leq p_i \leq 1$ for all $i$
- $\sum_i p_i = 1$ (the total probability equals 1)

```python
# Example: probability distribution for a six-sided die
outcomes = [1, 2, 3, 4, 5, 6]
probs = [1/6] * 6

print("Discrete uniform distribution (fair die):")
for o, p in zip(outcomes, probs):
    print(f"  P({o}) = {p:.4f}")

print(f"\nTotal probability: {sum(probs):.10f}")
print(f"Valid probability distribution: {abs(sum(probs) - 1) < 1e-10}")
```

### Expected Value and Variance

The **expected value** (mean) of a discrete variable is:
$$E[X] = \sum_i x_i \cdot p_i$$

The **variance** measures spread:
$$\text{Var}(X) = E[(X - \mu)^2] = \sum_i (x_i - \mu)^2 \cdot p_i$$

```python
# Expected value and variance for a die
prob_dist = list(zip(outcomes, probs))

# Expected value
expected_value = sum(x * p for x, p in prob_dist)
print(f"Expected value E[X] = {expected_value}")

# Variance
variance = sum((x - expected_value)**2 * p for x, p in prob_dist)
print(f"Variance Var(X) = {variance:.4f}")
print(f"Standard deviation σ = {variance**0.5:.4f}")
```

## 2.7 Complex Numbers

Now we arrive at the mathematical essence of quantum mechanics. **Complex numbers** are the natural language for probability amplitudes.

A complex number has a real part and an imaginary part: $z = a + ib$, where $i = \sqrt{-1}$ and $b \in \mathbb{R}$.

### Visualization on the Complex Plane

Using Euler's formula, we can write a complex number in **polar form**:
$$z = r e^{i\theta} = r(\cos\theta + i\sin\theta)$$

where $r = |z| = \sqrt{a^2 + b^2}$ is the magnitude and $\theta$ is the phase angle.

```python
# Complex number arithmetic
z1 = 3 + 4j
z2 = 1 - 2j

print(f"z1 = {z1}")
print(f"z2 = {z2}\n")

print(f"Addition (z1 + z2): {z1 + z2}")
print(f"Multiplication (z1 * z2): {z1 * z2}")

# Convert to polar form
r1 = abs(z1)
theta1 = cmath.phase(z1)
print(f"\nz1 in polar form: r = {r1:.4f}, θ = {theta1:.4f} rad")

# Verify conversion back: r·e^(iθ) = r·cos(θ) + i·r·sin(θ)
reconstructed = r1 * cmath.exp(1j * theta1)
print(f"Verification: {reconstructed} ≈ {z1}")
print(f"Match: {abs(reconstructed - z1) < 1e-14}")
```

### The Complex Conjugate

The complex conjugate of $z = a + ib$ is $z^* = a - ib$. It flips the sign of the imaginary part.

Key property: $z \cdot z^* = |z|^2$ — a real, non-negative number.

```python
# Complex conjugate
print(f"z1 = {z1}")
print(f"z1* (conjugate) = {z1.conjugate()}")
print(f"z1 · z1* = {z1 * z1.conjugate()}")
print(f"|z1|² = {z1.real**2 + z1.imag**2}")
print(f"In magnitude form: |z1| = {abs(z1)}")

# The result is purely real: 3² + 4² = 25
print(f"\nCheck: 3² + 4² = 9 + 16 = {3**2 + 4**2}")
```

### Born's Rule and Probability

**Born's Rule** (Max Born, 1926) states that the probability of finding a quantum system in a specific state is the magnitude squared of its probability amplitude:
$$P = |\psi|^2 = \psi \cdot \psi^*$$

```python
# Born's rule example
amplitude = (3 + 4j) / 5  # normalized so |amplitude|² = 1

probability = amplitude * amplitude.conjugate()

print(f"Probability amplitude: {amplitude}")
print(f"Probability = |amplitude|² = {probability.real:.4f}")
print(f"Magnitude: |amplitude| = {abs(amplitude):.4f}")
print(f"\nThe amplitude is normalized: |amplitude|² = 1.0 (as expected)")
```

## 2.8 Vectors and Matrices (Linear Algebra)

A quantum state is represented as a **vector** in a complex vector space (Hilbert space). Physical operations are represented by **matrices** that transform these vectors.

### Column Vectors and Ket Notation

Dirac's **bra-ket notation**:
- A **ket** $|\psi\rangle$ is a column vector
- A **bra** $\langle\psi|$ is the conjugate transpose (row vector)

The two fundamental qubit states:

```python
import numpy as np

# Define the computational basis states
ket_0 = np.array([[1],
                  [0]])
ket_1 = np.array([[0],
                  [1]])

print("State |0> = ", ket_0.T)
print("State |1> = ", ket_1.T)
print("\nWritten as column vectors:")
print("|0⟩ = [1, 0]^T")
print("|1⟩ = [0, 1]^T")
```

### Inner Products (Bra·Ket)

The inner product $\langle\phi|\psi\rangle$ tells us how "aligned" two states are. For orthogonal states, the inner product is zero.

```python
# The bra is the conjugate transpose of the ket
# In numpy: H is conjugate transpose (Hermitian transpose)
bra_0 = ket_0.T.conj()  # <0|
bra_1 = ket_1.T.conj()  # <1|

# Inner products
overlap_01 = bra_0 @ ket_1  # <0|1>
overlap_00 = bra_0 @ ket_0  # <0|0>
overlap_11 = bra_1 @ ket_1  # <1|1>

print(f"<0|0> = {overlap_00} (orthonormal: yes)")
print(f"<1|1> = {overlap_11} (orthonormal: yes)")
print(f"<0|1> = {overlap_01} (orthogonal: yes)")
```

### Vector Norm (Normalization)

A valid quantum state must have unit norm: $\langle\psi|\psi\rangle = 1$.

```python
# Check normalization of |0⟩ and |1⟩
norm_0 = np.sqrt(np.real((bra_0 @ ket_0).real))
norm_1 = np.sqrt(np.real((bra_1 @ ket_1).real))
print(f"||0⟩|| = {norm_0:.4f}")
print(f"||1⟩|| = {norm_1:.4f}")
print(f"Both normalized: {abs(norm_0 - 1) < 1e-10 and abs(norm_1 - 1) < 1e-10}")
```

### Superposition

A quantum system can exist in a combination of states:
$$|\psi\rangle = \alpha |0\rangle + \beta |1\rangle$$

where $\alpha$ and $\beta$ are complex probability amplitudes satisfying $|\alpha|^2 + |\beta|^2 = 1$.

```python
# Create a superposition state
alpha = 1 / np.sqrt(2)
beta = 1j / np.sqrt(2)  # beta has a phase (imaginary part)

# Verify normalization: |α|² + |β|² = 1
norm_sq = abs(alpha)**2 + abs(beta)**2
print(f"Superposition: |ψ⟩ = {alpha}|0⟩ + {beta}i|1⟩")
print(f"Normality check: |α|² + |β|² = {norm_sq:.4f}")

# The ket vector
ket_psi = alpha * ket_0 + beta * ket_1
print(f"\n|ψ> = {ket_psi.flatten()}")

# Probabilities of each outcome
p0 = abs(alpha)**2
p1 = abs(beta)**2
print(f"\nIf we measure:")
print(f"  P(0) = |α|² = {p0:.4f} → {p0 * 100:.0f}%")
print(f"  P(1) = |β|² = {p1:.4f} → {p1 * 100:.0f}%")
print(f"  Total = {p0 + p1:.4f}")
```

### Quantum Operators (Matrices)

Operations on states are represented by matrices. A common operator is the **Pauli-X matrix** (quantum NOT gate), which flips $|0\rangle \leftrightarrow |1\rangle$:
$$X = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}$$

```python
# Pauli-X matrix
X_gate = np.array([[0, 1],
                   [1, 0]])

# Apply X to |0>
X_on_0 = X_gate @ ket_0
print("X|0⟩ =", X_on_0.flatten())
print("Expected: |1⟩ = [0, 1]")
print(f"Match: {np.array_equal(np.round(X_on_0.flatten(), 10), ket_1.flatten())}")

# Apply X to |1>
X_on_1 = X_gate @ ket_1
print(f"\nX|1⟩ = {X_on_1.flatten()}")
print("Expected: |0⟩ = [1, 0]")
print(f"Match: {np.array_equal(np.round(X_on_1.flatten(), 10), ket_0.flatten())}")
```

### The Hadamard Gate

The **Hadamard gate** creates equal superpositions:
$$H = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 & 1 \\ 1 & -1 \end{pmatrix}$$

It maps $|0\rangle \to |\psi_{\text{equal}}\rangle$, and $|1\rangle \to \frac{|0\rangle - |1\rangle}{\sqrt{2}}$.

```python
# Hadamard gate
H = (1 / np.sqrt(2)) * np.array([[1, 1],
                                 [1, -1]])

H_on_0 = H @ ket_0
H_on_1 = H @ ket_1

print("H|0⟩ =", H_on_0.flatten())
print("H|1⟩ =", H_on_1.flatten())

# Verify both are normalized
norm_H0 = np.sqrt(np.real(H_on_0.T.conj() @ H_on_0))
norm_H1 = np.sqrt(np.real(H_on_1.T.conj() @ H_on_1))
print(f"\n||H|0⟩|| = {norm_H0:.4f}")
print(f"||H|1⟩|| = {norm_H1:.4f}")
```
