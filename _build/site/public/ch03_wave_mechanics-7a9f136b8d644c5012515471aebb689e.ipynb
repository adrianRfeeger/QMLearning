# 3. Wave Mechanics and the Schrödinger Equation

In 1926, Erwin Schrödinger formulated an equation that describes how the quantum state of a physical system changes over time. It is to quantum mechanics what Newton's second law ($F = ma$) is to classical mechanics.

## 3.1 The Time-Independent Schrödinger Equation

For systems where the energy doesn't change over time, we use the time-independent Schrödinger equation (in one dimension):

$$\hat{H} \psi(x) = E \psi(x)$$

Here:
- $\hat{H}$ is the **Hamiltonian operator**, representing the total energy (kinetic + potential) of the system.
- $\psi(x)$ is the **wave function**, which we are trying to solve for.
- $E$ is the total **energy** associated with that wave function.

## 3.2 The Particle in a 1D Box

Imagine a particle trapped in a one-dimensional box of length $L$. It can move freely inside the box (potential energy $V = 0$), but cannot escape (potential energy $V = \infty$ at the walls).

Because of these boundaries, the wave function must be zero at the walls (like a guitar string tied at both ends). This leads to **quantized** energy levels and wave functions. The allowed wave functions are standing waves:

$$\psi_n(x) = \sqrt{\frac{2}{L}} \sin\left(\frac{n\pi x}{L}\right)$$

where $n = 1, 2, 3, \dots$ is the principal quantum number. Let's visualize these wave functions!

```python
import numpy as np
import matplotlib.pyplot as plt

L = 1.0  # Length of the box
x = np.linspace(0, L, 500)

def psi(n, x, L):
    return np.sqrt(2/L) * np.sin(n * np.pi * x / L)

plt.figure(figsize=(10, 6))

# Plot the first three energy levels (n=1, 2, 3)
for n in [1, 2, 3]:
    plt.plot(x, psi(n, x, L), label=f'n={n}')

plt.title('Wave Functions for a Particle in a 1D Box')
plt.xlabel('Position x')
plt.ylabel('Wave Function $\psi_n(x)$')
plt.legend()
plt.grid(True)
plt.show()
```

## 3.3 Probability Density

The wave function $\psi(x)$ itself doesn't have a direct physical meaning. However, its magnitude squared, $|\psi(x)|^2$, represents the **probability density** of finding the particle at position $x$. Let's plot the probabilities for the same states.

```python
plt.figure(figsize=(10, 6))

for n in [1, 2, 3]:
    # Probability density is the square of the wave function
    prob_density = np.abs(psi(n, x, L))**2
    plt.plot(x, prob_density, label=f'n={n}')

plt.title('Probability Density for a Particle in a 1D Box')
plt.xlabel('Position x')
plt.ylabel('Probability Density $|\psi_n(x)|^2$')
plt.legend()
plt.grid(True)
plt.show()
```

Notice something strange? For $n=2$, the probability of finding the particle exactly in the middle of the box ($x=0.5$) is strictly **zero**! The particle can be on the left side, or the right side, but never in the exact middle. This is entirely non-classical behavior.
