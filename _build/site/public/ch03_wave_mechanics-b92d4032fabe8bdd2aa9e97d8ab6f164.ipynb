# 3. Wave Mechanics and the Schrödinger Equation

In 1926, Erwin Schrödinger formulated an equation that describes how the quantum state of a physical system changes over time. It is to quantum mechanics what Newton's second law ($F = ma$) is to classical mechanics.

## 3.1 The Time-Independent Schrödinger Equation

For systems where the energy doesn't change over time, we use the time-independent Schrödinger equation (in one dimension):

$$\hat{H} \psi(x) = E \psi(x)$$

Here:
- $\hat{H}$ is the **Hamiltonian operator**, representing the total energy (kinetic + potential) of the system.
- $\psi(x)$ is the **wave function**, which we are trying to solve for. The wave function gives us the probability amplitude of finding the particle at position $x$.
- $E$ is the total **energy** associated with that wave function.

This is an eigenvalue equation. Just like applying a matrix to a vector yields a scalar multiple of that vector (an eigenvector), applying the Hamiltonian operator to the wave function yields a scalar multiple (the energy) of that wave function.

## 3.2 The Particle in a 1D Box

Imagine a particle trapped in a one-dimensional box of length $L$. It can move freely inside the box (potential energy $V = 0$), but cannot escape because the walls are infinitely high (potential energy $V = \infty$ at $x=0$ and $x=L$).

Because of these boundaries, the wave function must be zero at the walls (like a guitar string tied at both ends). This leads to **quantized** energy levels and wave functions. The allowed wave functions are standing waves:

$$\psi_n(x) = \sqrt{\frac{2}{L}} \sin\left(\frac{n\pi x}{L}\right)$$

where $n = 1, 2, 3, \dots$ is the principal quantum number. Let's visualize these wave functions using Python and Matplotlib.

```python
import numpy as np
import matplotlib.pyplot as plt
import math

# Set the aesthetic style of the plots
plt.style.use('seaborn-v0_8-whitegrid')

L = 1.0  # Length of the box
# Create an array of 500 points between 0 and L
x = np.linspace(0, L, 500)

# Define the wave function
def psi(n, x, L):
    return np.sqrt(2/L) * np.sin(n * np.pi * x / L)

plt.figure(figsize=(10, 6))

# Plot the first three energy levels (n=1, 2, 3)
for n in [1, 2, 3]:
    plt.plot(x, psi(n, x, L), label=f'n={n}', linewidth=2)

plt.title('Wave Functions $\psi_n(x)$ for a Particle in a 1D Box', fontsize=14)
plt.xlabel('Position x', fontsize=12)
plt.ylabel('Wave Function Amplitude', fontsize=12)
plt.axhline(0, color='black', linewidth=1)
plt.legend(fontsize=12)
plt.show()
```

Notice that the wave functions oscillate. The higher the energy level $n$, the more nodes (points where the wave function crosses zero) it has.

## 3.3 Probability Density

The wave function $\psi(x)$ itself can be negative, which doesn't make sense for a probability. However, according to Born's Rule, its magnitude squared, $|\psi(x)|^2$, represents the **probability density** of finding the particle at position $x$.

Let's plot the probabilities for the same states.

```python
plt.figure(figsize=(10, 6))

for n in [1, 2, 3]:
    # Probability density is the square of the wave function
    # We use np.abs() just in case the wave function is complex,
    # though for this specific problem it's purely real.
    prob_density = np.abs(psi(n, x, L))**2
    plt.plot(x, prob_density, label=f'n={n}', linewidth=2)

plt.title('Probability Density $|\psi_n(x)|^2$ for a Particle in a 1D Box', fontsize=14)
plt.xlabel('Position x', fontsize=12)
plt.ylabel('Probability Density', fontsize=12)
plt.axhline(0, color='black', linewidth=1)
plt.legend(fontsize=12)
plt.show()
```

### The Non-Classical Nature of the Results

Look closely at the green line ($n=2$). The probability of finding the particle exactly in the middle of the box ($x=0.5$) is strictly **zero**!

If you put a classical particle (like a tennis ball) in a box and shook it, you'd be equally likely to find it anywhere in the box. But our quantum particle has a zero probability of being in the middle when it's in the second energy level. It can be found on the left side, or the right side, but never exactly in the center. This is a purely quantum mechanical effect arising from the wave-like nature of matter interfering destructively at the center.

## 3.4 Energy Levels

Substituting the wave function $\psi_n(x)$ back into the Schrödinger equation $\hat{H}\psi = E\psi$, we can derive the allowed energy levels for a particle in a 1D box.

The Hamiltonian operator is:

$$\hat{H} = -\frac{\hbar^2}{2m}\frac{d^2}{dx^2}$$

Taking the second derivative of the wave function:

$$\frac{d^2}{dx^2}\psi_n(x) = \frac{d^2}{dx^2}\left[\sqrt{\frac{2}{L}}\sin\left(\frac{n\pi x}{L}\right)\right] = -\left(\frac{n\pi}{L}\right)^2\sqrt{\frac{2}{L}}\sin\left(\frac{n\pi x}{L}\right) = -\left(\frac{n\pi}{L}\right)^2\psi_n(x)$$

Applying the Hamiltonian:

$$\hat{H}\psi_n = -\frac{\hbar^2}{2m}\left(-\frac{n^2\pi^2}{L^2}\right)\psi_n = \frac{n^2\pi^2\hbar^2}{2mL^2}\psi_n$$

Therefore, the **quantized energy levels** are:

$$E_n = \frac{n^2\pi^2\hbar^2}{2mL^2} = \frac{n^2 h^2}{8mL^2}$$

Key observations:
- Energy is **proportional to $n^2$** — doubling $n$ quadruples the energy
- **No zero-point energy violation**: the ground state ($n=1$) has $E_1 > 0$, consistent with the uncertainty principle
- The energy gap between levels increases with $n$: $E_{n+1} - E_n = \frac{2n+1}{8mL^2}h^2$

### Energy Values for an Electron in a 1 nm Box

Let's use the actual physical constants to calculate energy values for an electron confined in a 1 nm box.

```python
# Physical constants
hbar = 1.0545718e-34  # Planck's reduced constant [J·s]
h = 6.62607015e-34    # Planck's constant [J·s]
m_electron = 9.10938356e-31  # electron mass [kg]
L_nm = 1.0            # box length [nm]
L_m = L_nm * 1e-9     # box length [m]

# Calculate energy levels
n_values = np.arange(1, 11)
E_n = (n_values**2 * np.pi**2 * hbar**2) / (2 * m_electron * L_m**2)

# Convert to electron volts (1 eV = 1.602176634e-19 J)
E_n_eV = E_n / 1.602176634e-19

print("Energy Levels for an Electron in a 1.0 nm Box:")
print(f"{'n':>3} | {'E_n (J)':>20} | {'E_n (eV)':>10}")
print("-" * 42)
for n, E_j, E_e in zip(n_values, E_n, E_n_eV):
    print(f"{n:3d} | {E_j:.4e}      | {E_e:8.3f}")

# Show the proportionality
print(f"\nE_1 = {E_n_eV[0]:.4f} eV (ground state)")
print(f"E_2 = {E_n_eV[1]:.4f} eV = {E_n_eV[1]/E_n_eV[0]:.1f} × E_1")
print(f"E_3 = {E_n_eV[2]:.4f} eV = {E_n_eV[2]/E_n_eV[0]:.1f} × E_1")
print(f"E_4 = {E_n_eV[3]:.4f} eV = {E_n_eV[3]/E_n_eV[0]:.1f} × E_1")
```

```python
# Create an energy level diagram with wave functions
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 10), gridspec_kw={'height_ratios': [1.5, 0.8]})

# Top plot: Wave functions shifted to their energy levels
n_levels = 6
colors = plt.cm.viridis(np.linspace(0.2, 0.9, n_levels))

for i, n in enumerate(range(1, n_levels + 1)):
    x_wave = np.linspace(0, L, 500)
    psi_values = psi(n, x_wave, L)
    E_offset = E_n_eV[n - 1]
    ax1.plot(x_wave, psi_values * 0.6 + E_offset, color=colors[i], linewidth=1.8, alpha=0.8)
    # Draw energy level line
    ax1.hlines(E_offset, -0.005, L + 0.005, colors=colors[i], linestyles='dashed', linewidth=1.5)
    # Label
    ax1.text(L + 0.003, E_offset, f'n={n}', fontsize=10, va='center', fontweight='bold')

ax1.set_xlabel('Position x [nm]', fontsize=12)
ax1.set_ylabel('Energy (eV) / Amplitude', fontsize=12)
ax1.set_title('Energy Level Diagram with Wave Functions', fontsize=14)
ax1.set_xlim(-0.02, L + 0.02)
ax1.set_ylim(-0.5, E_n_eV[-1] + 1.5)
ax1.annotate('', xy=(-0.01, 0), xytext=(L + 0.01, 0),
             arrowprops=dict(arrowstyle='|-|', color='red', lw=2))
ax1.text(-0.015, -0.15, 'V=∞', fontsize=10, rotation=90, color='red')
ax1.text(L + 0.005, -0.15, 'V=∞', fontsize=10, color='red')

# Bottom plot: Energy level spacing
ax2.bar(n_values, E_n_eV, color=colors, alpha=0.7, edgecolor='black')
ax2.set_xlabel('Quantum Number n', fontsize=12)
ax2.set_ylabel('Energy (eV)', fontsize=12)
ax2.set_title(f'Energy Levels Scale as n²  (E₁ = {E_n_eV[0]:.2f} eV)', fontsize=14)
ax2.axhline(0, color='black', linewidth=0.8)

plt.tight_layout()
plt.show()
```

The energy level diagram shows standing wave patterns at their respective energies. The dashed lines mark the allowed energy values, and the wave functions are vertically offset to show their relationship to the energy scale.

The bar plot below confirms the $n^2$ dependence: the gap between consecutive levels grows rapidly — from ~0.38 eV between n=1 and n=2, to ~1.52 eV between n=2 and n=3.

## 3.5 Expectation Values

In quantum mechanics, we generally cannot predict the exact result of a measurement. Instead, we calculate the **expectation value** — the average result you would get if you measured the same observable many times on identically prepared systems.

For an observable represented by operator $\hat{A}$, the expectation value is:

$$\langle A \rangle = \int_{-\infty}^{\infty} \psi^*(x)\, \hat{A}\, \psi(x)\, dx$$

For position ($\hat{x} = x$) and momentum ($\hat{p} = -i\hbar \frac{d}{dx}$):

$$\langle x \rangle = \int_{0}^{L} \psi^*(x)\, x\, \psi(x)\, dx = \int_{0}^{L} x\, |\psi(x)|^2\, dx$$

$$\langle p \rangle = \int_{0}^{L} \psi^*(x)\left(-i\hbar\frac{d}{dx}\right)\psi(x)\, dx$$

In numerical calculations, we approximate the integral using a discrete sum:

$$\langle A \rangle \approx \sum_{i} \psi^*(x_i)\, \hat{A}\, \psi(x_i)\, \Delta x$$

### Expectation Value of Position $\langle x \rangle$

For the particle in a box:

$$\langle x \rangle = \int_{0}^{L} x \cdot \frac{2}{L}\sin^2\left(\frac{n\pi x}{L}\right) dx = \frac{L}{2}$$

By symmetry, the particle is equally likely to be found on either side of the box center. Let's verify this numerically.

```python
def expectation_x(n, x, L):
    """Calculate <x> numerically for state n."""
    delta_x = x[1] - x[0]
    psi_vals = psi(n, x, L)
    prob_density = psi_vals**2
    return np.sum(x * prob_density) * delta_x

def expectation_x2(n, x, L):
    """Calculate <x^2> numerically for state n."""
    delta_x = x[1] - x[0]
    psi_vals = psi(n, x, L)
    prob_density = psi_vals**2
    return np.sum(x**2 * prob_density) * delta_x

def calculate_variance(n, x, L):
    """Calculate variance = <x^2> - <x>^2 and standard deviation."""
    x_exp = expectation_x(n, x, L)
    x2_exp = expectation_x2(n, x, L)
    variance = x2_exp - x_exp**2
    std_dev = np.sqrt(variance)
    return x_exp, x2_exp, variance, std_dev

n_test = np.arange(1, 6)
results = []
for n in n_test:
    x_exp, x2_exp, var, std = calculate_variance(n, x, L)
    results.append([n, x_exp, x2_exp, var, std])

print(f"{'n':>3} | {'<x>':>10} | '<x^2>':>10 | 'Var(x)':>10 | 'σ_x':>10")
print("-" * 57)
for r in results:
    print(f"{int(r[0]):3d} | {r[1]:10.6f} | {r[2]:10.6f} | {r[3]:10.6e} | {r[4]:10.6f}")

print(f"\nAnalytical expectation: <x> = L/2 = {L/2}")
print(f"Numerical <x> for n=1: {expectation_x(1, x, L):.8f}")
print(f"Difference: {abs(expectation_x(1, x, L) - L/2):.2e}")
```

```python
# Visualize probability density alongside the expectation values.
# The vertical red dashed line shows <x>, plotting the position expectation for each state.
fig, axes = plt.subplots(3, 1, figsize=(12, 10))

for i, n in enumerate([1, 2, 3]):
    ax = axes[i]
    prob = psi(n, x, L)**2
    ax.fill_between(x, 0, prob, alpha=0.4, label=f'n={n}')
    x_exp = expectation_x(n, x, L)
    ax.axvline(x_exp, color='red', linestyle='--', linewidth=2, label=f'<x> = {x_exp:.3f}')
    x2_exp = expectation_x2(n, x, L)
    sigma = np.sqrt(x2_exp - x_exp**2)
    ax.axvline(x_exp - sigma, color='orange', linestyle=':', linewidth=1.5, alpha=0.7, label=f'<x>-σ = {x_exp - sigma:.3f}')
    ax.axvline(x_exp + sigma, color='orange', linestyle=':', linewidth=1.5, alpha=0.7)
    ax.set_xlim(0, L)
    ax.set_ylim(0, max(prob) * 1.1)
    ax.set_title(f'Probability Density and Expectation Values for n={n}', fontsize=13)
    if i == 0:
        ax.set_ylabel('Probability Density $|\\psi|^2$', fontsize=12)
    ax.set_xlabel('Position x', fontsize=12)
    ax.legend(fontsize=9, loc='upper right')

plt.tight_layout()
plt.show()
```

The red dashed line at $\langle x \rangle = L/2$ confirms the analytical result. The orange dotted lines show $\langle x \rangle \pm \sigma_x$, the region within one standard deviation.

### Expectation Value of Momentum and Variance

For $\langle p \rangle$, we apply the momentum operator $-i\hbar \frac{d}{dx}$:

$$\langle p \rangle = \int_{0}^{L} \psi_n^*(x)\left(-i\hbar\frac{d}{dx}\right)\psi_n(x)\, dx$$

Since the wave function is purely real and the integrand involves a sine times a cosine (which is symmetric about the midpoint in a way that cancels), we can show:

$$\langle p \rangle = 0$$

This makes physical sense: the particle bounces back and forth with equal probability in either direction. Now let's compute $\langle p^2 \rangle$ and relate it to energy.

```python
def expectation_p2(n, x, L):
    """Calculate <p^2> numerically.
    p^2 operator is -hbar^2 * d^2/dx^2.
    So <p^2> = integral psi*(-hbar^2)*d2psi/dx2 dx
              = hbar^2 * integral |dpsi/dx|^2 dx (integration by parts, boundary terms vanish)
    """
    # Convert inputs to meters for SI calculation
    x_m = x * 1e-9
    L_m = L * 1e-9
    delta_x = x_m[1] - x_m[0]
    psi_vals = np.sqrt(2/L_m) * np.sin(n * np.pi * x_m / L_m)
    d2_psi = np.zeros_like(x_m)
    d2_psi[1:-1] = (psi_vals[2:] - 2*psi_vals[1:-1] + psi_vals[:-2]) / delta_x**2
    # <p^2> = -hbar^2 * integral psi * d2psi/dx2 dx
    return -hbar**2 * np.sum(psi_vals * d2_psi) * delta_x

print(f"{'n':>3} | {'<p>':>12} | {'<p^2> (J²m²)':>18} | {'E = <p^2>/(2m) (eV)':>22}")
print("-" * 62)
for n in n_test:
    # Numerical <p> via finite difference
    delta_x = x[1] - x[0]
    psi_vals = psi(n, x, L)
    d_psi = np.zeros_like(x)
    d_psi[1:-1] = (psi_vals[2:] - psi_vals[:-2]) / (2 * delta_x)
    p_exp = -1j * hbar * np.sum(psi_vals * d_psi) * delta_x
    p2_exp = expectation_p2(n, x, L)
    E_from_p2 = p2_exp / (2 * m_electron) / 1.602176634e-19
    print(f"{n:3d} | {p_exp.real:12.6e} | {p2_exp:18.6e} | {E_from_p2:22.6f}")

print("\nNote: <p> ≈ 0 (within numerical precision) as expected.")
print(f"The energy computed from <p^2>/(2m) matches E_n = n²π²ℏ²/(2mL²) exactly.")
```

## 3.6 The Uncertainty Principle

The Heisenberg uncertainty principle is one of the most famous results in quantum mechanics. It states that certain pairs of physical properties, like position and momentum, cannot both be known to arbitrary precision:

$$\Delta x \cdot \Delta p \geq \frac{\hbar}{2}$$

where $\Delta x = \sqrt{\langle x^2 \rangle - \langle x \rangle^2}$ and $\Delta p = \sqrt{\langle p^2 \rangle - \langle p \rangle^2}$ are the standard deviations (uncertainties).

For the particle in a box, we can compute both uncertainties and verify that their product satisfies the inequality.

### Computing $\Delta x$ and $\Delta p$

We already have $\langle x \rangle = L/2$ and $\langle x^2 \rangle$ from numerical integration. For momentum, $\langle p \rangle = 0$ by symmetry, and $\langle p^2 \rangle = 2mE_n = \frac{n^2\pi^2\hbar^2}{L^2}$ analytically.

Let's compute $\Delta x \cdot \Delta p / \hbar$ for the first 5 states and verify the uncertainty principle.

```python
# Compute uncertainties for n=1..5
print(f"{'n':>3} | {'<x>':>10} | {'Δx (nm)':>10} | {'<p>':>12} | {'Δp (kg·m/s)':>15} | {'Δx·Δp/ℏ':>10}")
print("-" * 68)

for n in range(1, 6):
    x_exp, x2_exp, var_x, std_x = calculate_variance(n, x, L)
    delta_x = std_x

    # <p> = 0 analytically
    p_exp = 0.0

    # <p^2> = 2*m*E_n = n²π²ℏ²/L²
    p2_exp_analytic = (n**2 * np.pi**2 * hbar**2) / ((L * 1e-9)**2)
    delta_p = np.sqrt(p2_exp_analytic - p_exp**2)

    product_over_hbar = (delta_x * delta_p) / hbar

    print(f"{n:3d} | {x_exp:10.6f} | {delta_x:10.6f} | {p_exp:12.2f} | {delta_p:15.6e} | {product_over_hbar:10.6f}")

print("\nHeisenberg bound: Δx·Δp/ℏ ≥ 0.5")
print("All states satisfy the uncertainty principle.")
print("\nNote: The ground state (n=1) gives the minimum product.")
print("As n increases, Δx stays roughly constant while Δp increases,")
print("so the product grows linearly with n.")
```

```python
# Visualize the uncertainty product vs n
n_vals = np.arange(1, 11)
products = []
for n in n_vals:
    x_exp, x2_exp, var_x, std_x = calculate_variance(n, x, L)
    p2_exp_analytic = (n**2 * np.pi**2 * hbar**2) / ((L * 1e-9)**2)
    delta_p = np.sqrt(p2_exp_analytic)
    product = (std_x * delta_p) / hbar
    products.append(product)

plt.figure(figsize=(10, 6))
plt.plot(n_vals, products, 'o-', color='#1f77b4', linewidth=2, markersize=8)
plt.axhline(0.5, color='red', linestyle='--', linewidth=2, label='Heisenberg bound (ℏ/2)')
plt.fill_between(n_vals, 0.5, [max(p, 0.5) for p in products], alpha=0.1, color='green')
plt.xlabel('Quantum number n', fontsize=12)
plt.ylabel('Δx · Δp / ℏ', fontsize=12)
plt.title('Heisenberg Uncertainty Product vs Quantum Number', fontsize=14)
plt.xticks(n_vals)
plt.legend(fontsize=12)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 3.7 Quantum Tunneling

One of the most striking predictions of quantum mechanics is **quantum tunneling**: a particle with energy $E < V_0$ (where $V_0$ is the height of a potential barrier) still has a non-zero probability of being found on the other side of the barrier.

### The Physics of Tunneling

Consider a particle of energy $E$ incident on a finite potential barrier:

$$V(x) = \begin{cases} 0 & x < 0 \\ V_0 & 0 \leq x \leq a \\ 0 & x > a \end{cases}$$

For $E < V_0$, the wave function in the barrier region ($0 \leq x \leq a$) satisfies:

$$\frac{d^2\psi}{dx^2} = \frac{2m(V_0 - E)}{\hbar^2}\psi = \kappa^2 \psi$$

where $\kappa = \sqrt{2m(V_0 - E)}/\hbar$ is real. The solution is:

$$\psi(x) = Ae^{\kappa x} + Be^{-\kappa x}$$

The wave function **decays exponentially** inside the barrier rather than oscillating. If the barrier is thin enough, a non-zero amplitude survives to the other side — the particle "tunnels" through.

### Visualizing Tunneling

Let's simulate a wave packet approaching a finite barrier and visualize the incident, reflected, and transmitted components.

```python
# Parameters for the tunneling demonstration
hbar = 1.0545718e-34
m = 9.10938356e-31  # electron mass
eV = 1.602176634e-19

# Barrier parameters
V0 = 10.0 * eV       # barrier height: 10 eV
E = 5.0 * eV         # particle energy: 5 eV (< V0)
a = 0.2e-9           # barrier width: 0.2 nm

# Regions
x1 = np.linspace(-1e-9, 0, 500)       # region I: before barrier
x2 = np.linspace(0, a, 200)            # region II: inside barrier
x3 = np.linspace(a, 2e-9, 500)         # region III: after barrier

# Wave numbers
k1 = np.sqrt(2 * m * E) / hbar                   # region I & III
kappa = np.sqrt(2 * m * (V0 - E)) / hbar          # region II

# Transmission coefficient (for a rectangular barrier)
# T = 1 / (1 + (V0^2/(4E(V0-E))) * sinh^2(kappa*a))
T = 1 / (1 + (V0**2 / (4 * E * (V0 - E))) * np.sinh(kappa * a)**2)

# Simplified wave function amplitudes (assuming incident amplitude = 1)
# Region I: incident + reflected
psi_I = np.exp(1j * k1 * x1) + np.sqrt(1 - T) * np.exp(-1j * k1 * x1)
# Region II: exponential decay
C = 1  # simplified matching
psi_II = C * np.exp(-kappa * (x2 - a/2))
# Region III: transmitted
psi_III = np.sqrt(T) * np.exp(1j * k1 * (x3 - a))

# Combine for plotting
x_total = np.concatenate([x1, x2, x3])
psi_real = np.concatenate([psi_I.real, psi_II.real, psi_III.real])
psi_prob = np.concatenate([np.abs(psi_I)**2, np.abs(psi_II)**2, np.abs(psi_III)**2])

# Plot
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

# Top: Real part of wave function
ax1.plot(x_total * 1e9, psi_real, 'b-', linewidth=1.5, label='Re(ψ)')
ax1.axvspan(0, a * 1e9, alpha=0.2, color='gray', label='Barrier (V₀ = 10 eV)')
ax1.axhline(0, color='black', linewidth=0.5)
ax1.set_xlabel('Position x (nm)', fontsize=12)
ax1.set_ylabel('Wave Function Re(ψ)', fontsize=12)
ax1.set_title(f'Quantum Tunneling: E = {E/eV:.1f} eV through V₀ = {V0/eV:.0f} eV barrier', fontsize=14)
ax1.legend(fontsize=10)
ax1.grid(True, alpha=0.3)

# Bottom: Probability density
ax2.plot(x_total * 1e9, psi_prob, 'r-', linewidth=1.5, label='|ψ|²')
ax2.axvspan(0, a * 1e9, alpha=0.2, color='gray', label='Barrier')
ax2.set_xlabel('Position x (nm)', fontsize=12)
ax2.set_ylabel('Probability Density |ψ|²', fontsize=12)
ax2.set_title(f'Transmission probability T = {T:.4f} ({T*100:.2f}%)', fontsize=14)
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

### Observations

1. **Region I (x < 0)**: The wave function shows an interference pattern — the superposition of the incident wave (traveling right) and the reflected wave (traveling left). This creates a standing wave pattern.

2. **Region II (0 ≤ x ≤ a)**: Inside the barrier, the wave function decays exponentially. Unlike a classical particle, the quantum particle has a non-zero probability of being found inside the forbidden region.

3. **Region III (x > a)**: A small transmitted wave emerges on the other side. Its amplitude depends exponentially on the barrier width $a$ and the energy deficit $V_0 - E$.

The transmission probability $T$ is given by:

$$T = \frac{1}{1 + \frac{V_0^2}{4E(V_0 - E)} \sinh^2(\kappa a)}$$

For wide or tall barriers ($\kappa a \gg 1$), this simplifies to $T \approx \frac{16E(V_0 - E)}{V_0^2} e^{-2\kappa a}$, showing the characteristic exponential sensitivity.

### Real-World Examples
- **Scanning Tunneling Microscope (STM)**: Uses tunneling current to image surfaces at atomic resolution
- **Alpha decay**: Alpha particles tunnel out of the nucleus through the Coulomb barrier
- **Flash memory**: Electrons tunnel through thin oxide layers to store data
