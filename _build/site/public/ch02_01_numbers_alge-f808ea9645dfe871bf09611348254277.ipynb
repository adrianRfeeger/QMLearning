# 2. Mathematical Foundations

To truly understand quantum mechanics, we must speak its language. Unlike classical physics, which relies heavily on calculus to track the continuous motion of objects — like a ball rolling down a hill — quantum mechanics relies primarily on **linear algebra** and **complex numbers**.

> **Why does this matter?** In the quantum world, particles don't have definite properties until we measure them. Instead of asking "where is the electron?" we ask "what is the *probability* of finding the electron here?" Linear algebra and complex numbers give us the mathematical tools to describe this probabilistic world.

This chapter builds the required mathematical tools from the ground up — starting with arithmetic and algebra, progressing through trigonometry and functions, and culminating in the vector spaces and complex numbers that form the backbone of quantum theory.

## 2.1 Arithmetic and Number Systems

Before we build complex mathematics, let's review the number systems you've been building on since primary school. Each system extends the one before it, giving us new kinds of problems we can solve.

### Integers, Fractions, and Decimals

We begin with the most basic number systems.

- **Integers** ($\mathbb{Z} = \dots, -2, -1, 0, 1, 2, \dots$) are the counting numbers extended to include negatives. Think of them as describing gains and losses: if you have $5 and owe someone $3, you effectively have $5 + (-3) = 2$.

- **Rational numbers** ($\mathbb{Q}$) are any numbers expressible as a fraction $\frac{p}{q}$ where $p$ and $q$ are integers and $q \neq 0$. This includes all terminating decimals (like $0.25 = \frac{1}{4}$) and repeating decimals (like $0.333\dots = \frac{1}{3}$).

- **Irrational numbers** (covered next) are the ones that *can't* be written as fractions — and they turn out to be everywhere in nature.

> **Real-world connection:** The ratio of a circle's circumference to its diameter is $\pi$, an irrational number. Every clock, every wheel, every wave depends on this number — and it can never be written exactly as a fraction.

```python
# Basic arithmetic operations
a, b = 7, 3

print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} × {b} = {a * b}")
print(f"Division: {a} ÷ {b} = {a / b}")
print(f"Integer division: {a} // {b} = {a // b}")
print(f"Remainder: {a} % {b} = {a % b}")
print(f"Exponentiation: {a} ^ {b} = {a ** b}")

# Try changing the values of a and b to experiment!
# Notice the difference between / (regular division) and // (integer division):
#   7 / 3 = 2.333...  (the actual answer, as a decimal)
#   7 // 3 = 2        (just the whole number part)
#   7 % 3 = 1         (what's left over — 3 goes into 7 twice, with 1 remaining)
```

### Irrational Numbers and $\pi$

Not all numbers can be written as fractions. **Irrational numbers** like $\pi$ and $\sqrt{2}$ have non-repeating, non-terminating decimal expansions. No matter how many digits you calculate, the pattern never settles into a repeat.

Here's a fascinating fact: **irrational numbers are actually more common than rational ones**. If you pick a random number on the number line, the chances are overwhelming that it's irrational.

Why do we care about them in quantum mechanics? Because waves and oscillations — the very fabric of quantum behaviour — are deeply connected to circular geometry, which brings $\pi$ into every equation.

#### Visualising $\pi$

Imagine a circle with radius $r = 1$ (called a *unit circle*). Its circumference is $2\pi \times 1 = 2\pi \approx 6.283$. This constant ratio appears in:
- The wavelength of light
- The energy levels of electrons in atoms
- The diffraction patterns in the double-slit experiment

Similarly, $\sqrt{2}$ appears when we calculate the length of a diagonal across a square — and it pops up when we compute the magnitude of complex numbers (more on that shortly).

```python
import math

# The number pi
print(f"π ≈ {math.pi:.15f}")

# Square roots of non-perfect squares are irrational
print(f"√2 ≈ {math.sqrt(2):.15f}")
print(f"√3 ≈ {math.sqrt(3):.15f}")

# The imaginary unit: sqrt(-1), which we'll revisit shortly
print(f"√(-1) is the imaginary unit i (or j in Python)")
```

## 2.2 Exponents and Logarithms

Exponentials and logarithms appear everywhere in quantum mechanics — from wave amplitudes to probability decays. Understanding them intuitively is key.

### What are exponents?

An exponent tells you how many times to multiply a number by itself. For example:
- $2^3 = 2 \times 2 \times 2 = 8$ (we say "2 to the power of 3")
- $5^2 = 5 \times 5 = 25$ (we say "5 squared" — this comes from calculating the area of a square)
- $3^4 = 3 \times 3 \times 3 \times 3 = 81$

But exponents go beyond whole numbers. You might have seen:
- $4^{1/2} = \sqrt{4} = 2$ (a half-power means square root)
- $8^{1/3} = \sqrt[3]{8} = 2$ (a third-power means cube root)
- $2^{-1} = \frac{1}{2}$ (a negative power means "take the reciprocal")

These extensions all follow the same consistent rules, which is what makes mathematics so elegant.

### The exponent rules

The three most important exponent rules are:
$$a^m \cdot a^n = a^{m+n}, \quad \frac{a^m}{a^n} = a^{m-n}, \quad (a^m)^n = a^{m \cdot n}$$

> **Why do these rules work?** Consider $a^3 \cdot a^2$. This means $(a \times a \times a) \times (a \times a)$ — we're multiplying $a$ by itself a total of $3 + 2 = 5$ times. So $a^3 \cdot a^2 = a^5$. The other rules follow similarly.

### The natural exponential function

The **natural exponential function** $e^x$ (where $e \approx 2.718$, known as *Euler's number*) is the most important function in quantum mechanics. Why? Because it naturally arises from differential equations — the mathematical language of change.

> **Where does $e$ come from?** One way to understand $e$ is through compound interest. If you invest money at 100% interest compounded continuously (meaning the interest is added at every instant, not just once a year), your money grows by a factor of $e$. This continuous growth model appears everywhere in nature: population growth, radioactive decay, cooling of hot objects, and quantum wavefunctions.

```python
# Exponentials
base, exp1, exp2 = 2, 3, 5

print(f"{base}^{exp1} = {base ** exp1}")        # 2 × 2 × 2 = 8
print(f"{base}^{exp2} = {base ** exp2}")        # 2 × 2 × 2 × 2 × 2 = 32

# Verify exponent rule: a^m * a^n = a^(m+n)
# When we multiply two powers with the same base, we add the exponents
print(f"\n{base}^{exp1} × {base}^{exp2} = {base ** exp1 * base ** exp2}")  # 8 × 32 = 256
print(f"{base}^({exp1}+{exp2}) = {base ** (exp1 + exp2)}")       # 2^8 = 256 ✓

# This rule works because:
# 2^3 × 2^5 = (2×2×2) × (2×2×2×2×2) = 2^8  (we count 8 twos total)
```

### Logarithms

The natural logarithm $\ln(x)$ is the inverse of $e^x$, meaning $\ln(e^x) = x$. You can think of it as answering the question: "To what power must we raise $e$ to get $x$?"

If $e^3 \approx 20.085$, then $\ln(20.085) \approx 3$.

Logarithms are the mathematical equivalent of "undoing" an exponential. Just as subtraction undoes addition and division undoes multiplication, the logarithm undoes exponentiation.

> **Why logarithms in quantum mechanics?** Many quantum processes involve exponential growth or decay (such as radioactive decay or the probability amplitude of finding a particle in a potential well). When we want to extract information from these exponential relationships, logarithms are our primary tool.

#### Common logarithmic rules
$$\ln(ab) = \ln(a) + \ln(b) \qquad \ln\!\left(\frac{a}{b}\right) = \ln(a) - \ln(b) \qquad \ln(a^n) = n\,\ln(a)$$

These rules are incredibly powerful — they let us convert multiplication into addition, which is why logarithmic scales (like the Richter scale for earthquakes and decibels for sound) are so useful in science.

```python
# Example: finding what power of e gives us x
x = math.exp(3)   # e^3 ≈ 20.086
log_x = math.log(x)  # natural log: "to what power must we raise e to get x?"

print(f"e^3 ≈ {x:.6f}")         # e cubed is about 20.086
print(f"ln(e^3) = {log_x}")     # taking the natural log undoes the exponential: answer is 3
print(f"So ln and e^ cancel each other out: {abs(log_x - 3) < 1e-10}")  # YES!

# Log rules: ln(ab) = ln(a) + ln(b)
# This is the logarithmic equivalent of the exponent rule: a^m × a^n = a^(m+n)
a, b = 5, 7
print(f"\nln({a} × {b}) = ln({a * b}) = {math.log(a * b):.6f}")    # ln(35)
print(f"ln({a}) + ln({b}) = {math.log(a) + math.log(b):.6f}")      # ln(5) + ln(7)
# These are equal! Logarithms turn multiplication into addition.
```

## 2.3 Algebra and Equations

Algebra lets us solve for unknown quantities by manipulating equations while keeping them balanced. Think of an equation like a set of scales: whatever you do to one side, you must do to the other. In quantum mechanics, we solve equations constantly — to find the energy levels of electrons, the probabilities of different outcomes, and the behaviour of waves.

### Solving Linear Equations

The simplest equation is linear: it involves only first powers of the unknown. Given $ax + b = c$, we solve for $x$:
1. Subtract $b$ from both sides: $ax = c - b$
2. Divide by $a$: $x = \frac{c - b}{a}$

That's it! The key is to isolate $x$ by performing the *inverse* operations in the reverse order that they were applied.

```python
# Solve ax + b = c for x: x = (c - b) / a
a, b, c = 3, 7, 16

# Step-by-step:
#   3x + 7 = 16
#   3x     = 16 - 7     (subtract 7 from both sides)
#   3x     = 9
#   x      = 9 / 3      (divide both sides by 3)
#   x      = 3

x = (c - b) / a
print(f"{a}x + {b} = {c} → x = (c - b) / a = ({c} - {b}) / {a} = {x}")

# Verify by plugging x back into the original equation
print(f"Check: {a}({x}) + {b} = {a * x + b}")  # Should equal 16
```

### Quadratic Equations

Quadratic equations are slightly more challenging because they involve $x^2$. They appear everywhere — from calculating the trajectory of a ball to determining the energy levels in quantum systems.

Given $ax^2 + bx + c = 0$, the **quadratic formula** gives:
$$x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$$

> **Where does this formula come from?** It's derived by a technique called "completing the square" — a brilliant algebraic trick. We add and subtract a carefully chosen number to turn the left side into a perfect square. While the derivation is interesting, what matters most is understanding the *parts* of the formula:
> - The $-b$ shifts the result based on the linear term.
> - The $\pm$ means there are **two** solutions (unless they happen to be the same).
> - The square root introduces the possibility of complex numbers when the discriminant is negative.
> - The $2a$ at the bottom normalises the result.

#### The Discriminant: $\Delta = b^2 - 4ac$

The discriminant tells us what *kind* of solutions we'll get:
- $\Delta > 0$: Two distinct **real** solutions (the parabola crosses the x-axis twice).
- $\Delta = 0$: One **repeated** solution (the parabola just touches the x-axis).
- $\Delta < 0$: Two **complex** solutions (the parabola never crosses the x-axis).

The third case is particularly important in quantum mechanics, because complex numbers are the language of quantum amplitudes.

```python
# Solve ax^2 + bx + c = 0
a, b, c = 1, -5, 6  # x^2 - 5x + 6 = 0, roots at x=2 and x=3

discriminant = b**2 - 4*a*c
root1 = (-b + math.sqrt(discriminant)) / (2*a)
root2 = (-b - math.sqrt(discriminant)) / (2*a)

print(f"{a}x² + ({b})x + {c} = 0")
print(f"Discriminant: Δ = b² - 4ac = {discriminant}")
print(f"Root 1: x = {root1}")
print(f"Root 2: x = {root2}")

# Step-by-step:
#   1) Calculate the discriminant: Δ = (-5)² - 4(1)(6) = 25 - 24 = 1
#   2) Take the square root: √1 = 1
#   3) Apply the formula: x = (5 ± 1) / 2
#   4) Root 1 = (5 + 1) / 2 = 3, Root 2 = (5 - 1) / 2 = 2

# Verify by plugging roots back into the original equation
print(f"\nCheck root 1: {a}*({root1})² + {b}*({root1}) + {c} = {a*root1**2 + b*root1 + c}")
print(f"Check root 2: {a}*({root2})² + {b}*({root2}) + {c} = {a*root2**2 + b*root2 + c}")
# Both should equal 0 — confirming our answers are correct!

# Try changing a, b, c to try different quadratic equations!
# What happens when the discriminant is negative?
```

### What if the discriminant is negative?

When $b^2 - 4ac < 0$, the square root involves **negative numbers**, which leads us to **complex numbers** — the language of quantum amplitudes.

> **A quick puzzle:** What number, when multiplied by itself, gives $-1$? No real number satisfies this, because a positive times a positive is positive, and a negative times a negative is also positive. This is why mathematicians invented the **imaginary unit**, denoted $i$ (or $j$ in engineering and Python), where $i = \sqrt{-1}$.

With this single invention, we can now take the square root of *any* number. And remarkably, this seemingly abstract concept turns out to be essential for describing how quantum particles behave.

```python
# Solve x^2 + 4 = 0, which has no real solution
a2, b2, c2 = 1, 0, 4  # x^2 + 4 = 0

discriminant2 = b2**2 - 4*a2*c2
print(f"{a2}x² + {b2}x + {c2} = 0")
print(f"Discriminant: Δ = {discriminant2} (negative → complex roots)")

# Step-by-step:
#   1) Δ = 0² - 4(1)(4) = -16
#   2) √(-16) = √(16) × √(-1) = 4i
#   3) x = (0 ± 4i) / 2 = ±2i
# So the two roots are x = 2i and x = -2i

# In Python, we create complex numbers using the 'j' suffix (engineering notation)
root1 = complex(0, 2)  # 0 + 2i
root2 = complex(0, -2) # 0 - 2i
print(f"Roots: x = {root1} and x = {root2}")

# Verify: (2i)² = 4i² = 4(-1) = -4, so x² + 4 = -4 + 4 = 0 ✓
print(f"\nVerification: (2i)² = {(2j)**2}")
print(f"So (2i)² + 4 = {(2j)**2 + 4}")  # Should equal 0
```
