## 2.6 Probability and Statistics Basics

Quantum mechanics is inherently *probabilistic*. This is not because we lack good enough measuring instruments — it is a fundamental feature of nature itself. We cannot predict with certainty what a single measurement will give us. All we can predict are the *probabilities* of different outcomes.

> **Why is this so strange?** In our everyday experience, probabilities usually reflect our ignorance. If I flip a coin and hide it under my hand, I say "there's a 50% chance it's heads" — but the coin is *already* either heads or tails; I just don't know which. In quantum mechanics, however, the particle is genuinely in a *superposition* of states until measured. The outcome is not determined until the moment of measurement. This is what Einstein famously struggled with when he said "God does not play dice."

### Discrete Probability

For a discrete set of outcomes, if outcome $i$ has probability $p_i$, then two rules must be satisfied:
- $0 \leq p_i \leq 1$ for all $i$ — probabilities are never negative and never exceed 100%.
- $\sum_i p_i = 1$ — the total probability of *something* happening must be exactly 1 (or 100%).

#### A concrete example: rolling a die

A fair six-sided die has six possible outcomes (1, 2, 3, 4, 5, 6), each with probability $\frac{1}{6}$. If we roll the die many times, we expect each number to appear roughly one sixth of the time. But on any single roll, we cannot predict the outcome.

> **Key insight:** Even though we cannot predict individual outcomes, probability theory lets us make *very* accurate predictions about averages and distributions over many trials. This is enough for quantum mechanics to make incredibly precise predictions.

#### What if the die is not fair?

Suppose we have a loaded die where the probabilities are:
- P(1) = 0.1, P(2) = 0.1, P(3) = 0.1, P(4) = 0.1, P(5) = 0.1, P(6) = 0.5

The total still adds up to 1, but now we're much more likely to roll a 6. In quantum mechanics, the probabilities are often *not* equal — and calculating them is one of our main goals.

```python
# Example: probability distribution for a six-sided die
outcomes = [1, 2, 3, 4, 5, 6]
probs = [1/6] * 6

print("Discrete uniform distribution (fair die):")
for o, p in zip(outcomes, probs):
    print(f"  P({o}) = {p:.4f}")

print(f"\nTotal probability: {sum(probs):.10f}")
print(f"Valid probability distribution: {abs(sum(probs) - 1) < 1e-10}")
```

### Expected Value and Variance

Once we have a probability distribution, we can calculate summary statistics that tell us about the *typical* behaviour of the system.

The **expected value** (or mean) of a discrete random variable is the weighted average of all possible values, where the weights are the probabilities:
$$E[X] = \sum_i x_i \cdot p_i$$

Think of it as what you'd expect to get if you repeated the experiment many times and took the average. For a fair die, every outcome is equally likely ($\frac{1}{6}$), so the expected value is simply $\frac{1+2+3+4+5+6}{6} = 3.5$.

> **Note:** The expected value doesn't have to be one of the possible outcomes! You can never roll a 3.5 on a die, but it is still the long-run average.

The **variance** measures how spread out the outcomes are from the expected value:
$$\text{Var}(X) = E[(X - \mu)^2] = \sum_i (x_i - \mu)^2 \cdot p_i$$
where $\mu = E[X]$ is the mean.

A small variance means outcomes cluster tightly around the mean; a large variance means they are more spread out. The **standard deviation** ($\sigma = \sqrt{\text{Var}(X)}$) is simply the square root of the variance, expressed in the same units as the original values.

> **Why variance matters in quantum mechanics:** The Heisenberg uncertainty principle is fundamentally about the variance (or standard deviation) of position and momentum measurements. Even if you prepare identical quantum systems in identical states, the outcomes will vary — and the uncertainty principle puts a fundamental limit on how small the product of these variances can be.

```python
# Expected value and variance for a die
prob_dist = list(zip(outcomes, probs))

# Expected value
expected_value = sum(x * p for x, p in prob_dist)
print(f"Expected value E[X] = {expected_value}")

# Variance
variance = sum((x - expected_value)**2 * p for x, p in prob_dist)
print(f"Variance Var(X) = {variance:.4f}")
print(f"Standard deviation σ = {variance**0.5:.4f}")
```

## 2.7 Complex Numbers

Now we arrive at the mathematical essence of quantum mechanics. **Complex numbers** are not just a mathematical curiosity — they are the natural language for probability amplitudes, which are at the heart of quantum theory.

A complex number has a **real part** and an **imaginary part**:
$$z = a + ib$$
where $i = \sqrt{-1}$ is the imaginary unit and $a, b \in \mathbb{R}$ are real numbers.

> **A note on terminology:** Many people think of the "imaginary" part as something made up or unreal. But complex numbers are *very* real — they describe physical phenomena with incredible precision. The name is a historical accident (introduced by Descartes as a derogatory term), and a better name would be "extended numbers" since they extend the real numbers in a perfectly sensible way.

#### Visualising complex numbers: the Complex Plane

Real numbers live on a one-dimensional number line. Complex numbers live on a **two-dimensional plane** called the *complex plane*:
- The horizontal axis represents the **real part** ($a$).
- The vertical axis represents the **imaginary part** ($b$).

So the complex number $3 + 4i$ is the point $(3, 4)$ on this plane. This visualisation is incredibly helpful for understanding operations like multiplication and conjugation.

#### Polar form: magnitude and phase

Using Euler's formula, we can write a complex number in **polar form**:
$$z = r e^{i\theta} = r(\cos\theta + i\sin\theta)$$

where:
- $r = |z| = \sqrt{a^2 + b^2}$ is the **magnitude** (the distance from the origin to the point). This is also called the *modulus*.
- $\theta$ is the **phase angle** (the angle the line from the origin makes with the positive real axis, measured in radians). This is also called the *argument*.

> **Why two forms?** The Cartesian form ($a + ib$) is convenient for addition and subtraction. The polar form ($re^{i\theta}$) is convenient for multiplication, division, and powers. In quantum mechanics, the phase is crucial — it determines how quantum states interfere with each other.

```python
import cmath
# Complex number arithmetic
z1 = 3 + 4j
z2 = 1 - 2j

print(f"z1 = {z1}")
print(f"z2 = {z2}\n")

print(f"Addition (z1 + z2): {z1 + z2}")
print(f"Multiplication (z1 * z2): {z1 * z2}")

# Convert to polar form
r1 = abs(z1)
theta1 = cmath.phase(z1)
print(f"\nz1 in polar form: r = {r1:.4f}, θ = {theta1:.4f} rad")

# Verify conversion back: r·e^(iθ) = r·cos(θ) + i·r·sin(θ)
reconstructed = r1 * cmath.exp(1j * theta1)
print(f"Verification: {reconstructed} ≈ {z1}")
print(f"Match: {abs(reconstructed - z1) < 1e-14}")
```

### The Complex Conjugate

The complex conjugate of $z = a + ib$ is $z^* = a - ib$. Geometrically, it is the reflection of $z$ across the real axis (the horizontal axis) on the complex plane.

Key property: multiplying a complex number by its conjugate gives a purely real, non-negative number:
$$z \cdot z^* = (a + ib)(a - ib) = a^2 - (ib)^2 = a^2 + b^2 = |z|^2$$

This is incredibly useful because it lets us "extract" the magnitude of a complex number. Notice that this is the same as the Pythagorean theorem: the magnitude squared is the sum of the squares of the real and imaginary parts.

> **Why the conjugate matters in quantum mechanics:** When we calculate the probability of a quantum outcome, we multiply the amplitude by its complex conjugate (this is Born's rule, explained below). The conjugate ensures the result is always a real, non-negative number — because probabilities must be real and non-negative, never imaginary or negative.

```python
# Complex conjugate
print(f"z1 = {z1}")
print(f"z1* (conjugate) = {z1.conjugate()}")
print(f"z1 · z1* = {z1 * z1.conjugate()}")
print(f"|z1|² = {z1.real**2 + z1.imag**2}")
print(f"In magnitude form: |z1| = {abs(z1)}")

# The result is purely real: 3² + 4² = 25
print(f"\nCheck: 3² + 4² = 9 + 16 = {3**2 + 4**2}")
```

### Born's Rule and Probability

**Born's Rule** (Max Born, 1926) states that the probability of finding a quantum system in a specific state is the magnitude squared of its probability amplitude:
$$P = |\psi|^2 = \psi \cdot \psi^*$$

> **The big idea:** In quantum mechanics, the state of a system is described by a *complex number* called a probability amplitude (usually denoted $\psi$). You cannot observe this amplitude directly — it is not a physical quantity you can measure with an instrument. However, the *probability* of observing a particular outcome is the squared magnitude of this amplitude.

Let's break down why this is so profound:

1. **Amplitudes can interfere.** Because amplitudes are complex numbers, they can have different phases. When two amplitudes add up, their phases matter: if the phases align (both pointing in roughly the same direction), they *constructively* interfere and the probability is larger. If the phases oppose (pointing in opposite directions), they *destructively* interfere and the probability is smaller. This is the mathematical mechanism behind the double-slit interference pattern.

2. **Probabilities are always real and non-negative.** Even though amplitudes are complex, the probability $|\psi|^2 = \psi \cdot \psi^*$ is always a real, non-negative number — exactly what we need for a physical probability.

3. **The amplitude carries phase information.** Two states with the same magnitude but different phases are physically distinct. This phase difference is what makes quantum mechanics different from classical probability theory.

> **Historical note:** Max Born developed this rule while trying to understand the meaning of the wavefunction. He received the Nobel Prize in Physics in 1954 largely for this insight. Before Born, even Einstein and Schrödinger thought the wavefunction described a physical charge or mass density. Born realised it was about *probability* — a revolutionary idea at the time.

#### Normalisation condition

For a single particle, the total probability of finding it *somewhere* must be 1 (certainty). This means:
$$\sum_i |\psi_i|^2 = 1$$

This is why we say quantum states must be "normalised" — the sum of all probabilities must equal 1. It's the same mathematical structure as the Pythagorean identity: $\sin^2\theta + \cos^2\theta = 1$.

```python
# Born's rule example
amplitude = (3 + 4j) / 5  # normalized so |amplitude|² = 1

probability = amplitude * amplitude.conjugate()

print(f"Probability amplitude: {amplitude}")
print(f"Probability = |amplitude|² = {probability.real:.4f}")
print(f"Magnitude: |amplitude| = {abs(amplitude):.4f}")
print(f"\nThe amplitude is normalized: |amplitude|² = 1.0 (as expected)")
```
