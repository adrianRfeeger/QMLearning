# 2. Mathematical Foundations

To understand quantum mechanics, we need to learn the language it is written in. The two most important mathematical concepts are **complex numbers** and **linear algebra** (vectors and matrices).

## 2.1 Complex Numbers

In classical mechanics, we describe probabilities and physical quantities using real numbers (like $1.5$, $-3$, or $\pi$). In quantum mechanics, the fundamental quantities (probability amplitudes) are described by **complex numbers**.

A complex number has a real part and an imaginary part, usually written as $c = a + i b$, where $i = \sqrt{-1}$.

Let's see how we can work with complex numbers in Python:

```python
# In Python, the imaginary unit 'i' is represented by 'j'
c1 = 3 + 4j
c2 = 1 - 2j

# We can add, subtract, and multiply them
print("c1 + c2 =", c1 + c2)
print("c1 * c2 =", c1 * c2)

# The complex conjugate (changing the sign of the imaginary part)
print("Complex conjugate of c1:", c1.conjugate())

# The magnitude squared |c1|^2 is obtained by multiplying a number by its conjugate
magnitude_squared = c1 * c1.conjugate()
print("Magnitude squared of c1:", magnitude_squared.real)
```

In quantum mechanics, the probability of an event is given by the magnitude squared of the complex probability amplitude.

## 2.2 Vectors and Matrices (Linear Algebra)

Quantum states are often represented as **vectors**, and physical operations (like measuring the state or evolving it in time) are represented by **matrices**.

### Dirac Notation
Physicists use a special notation called **Dirac notation** (or bra-ket notation) for vectors:
- A column vector is called a **ket**: $|\psi\rangle$
- A row vector (the complex conjugate transpose of the ket) is called a **bra**: $\langle\psi|$

Let's represent a simple 2D quantum state (a qubit) using `numpy`.

```python
import numpy as np

# A ket vector representing a state (e.g. state 0)
ket_0 = np.array([[1],
                  [0]])

# A ket vector representing another state (e.g. state 1)
ket_1 = np.array([[0],
                  [1]])

# A superposition state: (1/sqrt(2)) * (|0> + |1>)
ket_psi = (1 / np.sqrt(2)) * (ket_0 + ket_1)
print("Superposition state |psi>:\n", ket_psi)
```

### Matrices as Operators
Operators change one state into another. We multiply the matrix by the vector to get the new state. One famous quantum operator is the **Pauli-X matrix** (often called the bit-flip operator because it turns a 0 into a 1 and vice versa).

```python
# The Pauli-X matrix
X_gate = np.array([[0, 1],
                   [1, 0]])

# Apply X gate to ket_0 using matrix multiplication (@ in numpy)
new_state = X_gate @ ket_0
print("Result of applying X to |0>:\n", new_state)
```

Notice how the vector `[1, 0]` flipped to `[0, 1]`. That's the core of how quantum computation and quantum states evolve mathematically!
