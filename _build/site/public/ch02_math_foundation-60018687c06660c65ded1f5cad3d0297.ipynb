# 2. Mathematical Foundations

To truly understand quantum mechanics, we must speak its language. Unlike classical mechanics, which relies heavily on calculus to track the continuous motion of objects, quantum mechanics relies primarily on **linear algebra** and **complex numbers**.

## 2.1 Complex Numbers

In classical physics, probabilities are always real numbers between 0 and 1. If you flip a coin, the probability of getting heads is $0.5$.

In quantum mechanics, before we calculate a probability, we work with a **probability amplitude**, which is a complex number. A complex number has two parts: a real part and an imaginary part, written as $z = a + i b$, where $i = \sqrt{-1}$.

Let's explore complex numbers using Python. In Python, the imaginary unit is denoted by `j` instead of `i`.

```python
# Define two complex numbers
z1 = 3 + 4j
z2 = 1 - 2j

print(f"z1 = {z1}")
print(f"z2 = {z2}\n")

# Basic arithmetic
print(f"Addition (z1 + z2): {z1 + z2}")
print(f"Multiplication (z1 * z2): {z1 * z2}\n")

# The complex conjugate flips the sign of the imaginary part
# If z = a + ib, then its conjugate z* = a - ib
print(f"Complex conjugate of z1: {z1.conjugate()}")
```

### Born's Rule and Probability

Why do we care about complex numbers? Because of **Born's Rule**. Formulated by Max Born in 1926, it states that the probability of finding a quantum system in a specific state is equal to the **magnitude squared** of its probability amplitude.

The magnitude squared of a complex number $z = a + ib$ is calculated by multiplying it by its complex conjugate $z^* = a - ib$:

$$|z|^2 = z \cdot z^* = (a + ib)(a - ib) = a^2 - i^2b^2 = a^2 + b^2$$

Let's calculate this for our number `z1` (where $a=3$ and $b=4$).

```python
# Calculate magnitude squared manually
magnitude_squared = z1 * z1.conjugate()

# Note: The result of this multiplication mathematically is purely real,
# but Python still stores it as a complex number with a 0j imaginary part.
# We extract the real part to get the final probability value.
print(f"Magnitude squared of z1 (|z1|^2): {magnitude_squared.real}")

# Notice that 3^2 + 4^2 = 9 + 16 = 25. The math checks out!
```

## 2.2 Vectors and Matrices (Linear Algebra)

A quantum state is represented as a **vector** living in an abstract vector space (specifically, a Hilbert space). When a physical process happens—like a particle moving or a laser interacting with an atom—we mathematically represent that action as a **matrix** multiplying our state vector.

### Dirac Notation (Bra-Ket Notation)

Physicist Paul Dirac introduced a very convenient notation for quantum states:
- A column vector representing a state is called a **ket**: $|\psi\rangle$
- A row vector (the complex conjugate transpose of the ket) is called a **bra**: $\langle\psi|$

Let's define the two fundamental states of a quantum bit (qubit): $|0\rangle$ and $|1\rangle$. We can represent these as column vectors using the `numpy` library.

```python
import numpy as np

# Define ket_0 as a 2x1 column vector
ket_0 = np.array([[1],
                  [0]])

# Define ket_1 as a 2x1 column vector
ket_1 = np.array([[0],
                  [1]])

print("State |0>:\n", ket_0)
print("\nState |1>:\n", ket_1)
```

### Superposition

A quantum system can exist in a combination of states simultaneously. This is called **superposition**. We simply add the vectors together, multiplied by their complex probability amplitudes.

Let's create an equal superposition state:
$$|\psi\rangle = \frac{1}{\sqrt{2}} |0\rangle + \frac{1}{\sqrt{2}} |1\rangle$$

```python
# Create the superposition state
amplitude = 1 / np.sqrt(2)
ket_psi = amplitude * ket_0 + amplitude * ket_1

print("Superposition state |psi>:\n", ket_psi)

# If we measure this state, the probability of finding it in |0> is the magnitude squared of its amplitude:
prob_0 = amplitude**2
print(f"\nProbability of measuring 0: {prob_0:.2f} (or {prob_0 * 100:.0f}%)")
```

### Quantum Operators (Matrices)

Operations on states are represented by matrices. A very common operator in quantum computing is the **Pauli-X matrix** (or quantum NOT gate). It flips $|0\rangle$ to $|1\rangle$, and $|1\rangle$ to $|0\rangle$.

$$X = \begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}$$

```python
# Define the Pauli-X matrix
X_gate = np.array([[0, 1],
                   [1, 0]])

# Apply the operator to state |0> using matrix multiplication (@ in Python)
new_state = X_gate @ ket_0

print("Original state |0>:\n", ket_0)
print("\nState after applying X gate:\n", new_state)

# Check if the new state is exactly equal to |1>
is_equal = np.array_equal(new_state, ket_1)
print(f"\nDoes X|0> equal |1>? {is_equal}")
```
