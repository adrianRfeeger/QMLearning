# 2. Mathematical Foundations

To truly understand quantum mechanics, we must speak its language. Unlike classical physics, which relies heavily on calculus to track the continuous motion of objects — like a ball rolling down a hill — quantum mechanics relies primarily on **linear algebra** and **complex numbers**.

> **Why does this matter?** In the quantum world, particles don't have definite properties until we measure them. Instead of asking "where is the electron?" we ask "what is the *probability* of finding the electron here?" Linear algebra and complex numbers give us the mathematical tools to describe this probabilistic world.

This chapter builds the required mathematical tools from the ground up — starting with arithmetic and algebra, progressing through trigonometry and functions, and culminating in the vector spaces and complex numbers that form the backbone of quantum theory.

## 2.1 Arithmetic and Number Systems

Before we build complex mathematics, let's review the number systems you've been building on since primary school. Each system extends the one before it, giving us new kinds of problems we can solve.

### Integers, Fractions, and Decimals

We begin with the most basic number systems.

- **Integers** ($\mathbb{Z} = \dots, -2, -1, 0, 1, 2, \dots$) are the counting numbers extended to include negatives. Think of them as describing gains and losses: if you have $5 and owe someone $3, you effectively have $5 + (-3) = 2$.

- **Rational numbers** ($\mathbb{Q}$) are any numbers expressible as a fraction $\frac{p}{q}$ where $p$ and $q$ are integers and $q \neq 0$. This includes all terminating decimals (like $0.25 = \frac{1}{4}$) and repeating decimals (like $0.333\dots = \frac{1}{3}$).

- **Irrational numbers** (covered next) are the ones that *can't* be written as fractions — and they turn out to be everywhere in nature.

> **Real-world connection:** The ratio of a circle's circumference to its diameter is $\pi$, an irrational number. Every clock, every wheel, every wave depends on this number — and it can never be written exactly as a fraction.

```python
# Basic arithmetic operations
a, b = 7, 3

print(f"Addition: {a} + {b} = {a + b}")
print(f"Subtraction: {a} - {b} = {a - b}")
print(f"Multiplication: {a} × {b} = {a * b}")
print(f"Division: {a} ÷ {b} = {a / b}")
print(f"Integer division: {a} // {b} = {a // b}")
print(f"Remainder: {a} % {b} = {a % b}")
print(f"Exponentiation: {a} ^ {b} = {a ** b}")

# Try changing the values of a and b to experiment!
# Notice the difference between / (regular division) and // (integer division):
#   7 / 3 = 2.333...  (the actual answer, as a decimal)
#   7 // 3 = 2        (just the whole number part)
#   7 % 3 = 1         (what's left over — 3 goes into 7 twice, with 1 remaining)
```

### Irrational Numbers and $\pi$

Not all numbers can be written as fractions. **Irrational numbers** like $\pi$ and $\sqrt{2}$ have non-repeating, non-terminating decimal expansions. No matter how many digits you calculate, the pattern never settles into a repeat.

Here's a fascinating fact: **irrational numbers are actually more common than rational ones**. If you pick a random number on the number line, the chances are overwhelming that it's irrational.

Why do we care about them in quantum mechanics? Because waves and oscillations — the very fabric of quantum behaviour — are deeply connected to circular geometry, which brings $\pi$ into every equation.

#### Visualising $\pi$

Imagine a circle with radius $r = 1$ (called a *unit circle*). Its circumference is $2\pi \times 1 = 2\pi \approx 6.283$. This constant ratio appears in:
- The wavelength of light
- The energy levels of electrons in atoms
- The diffraction patterns in the double-slit experiment

Similarly, $\sqrt{2}$ appears when we calculate the length of a diagonal across a square — and it pops up when we compute the magnitude of complex numbers (more on that shortly).

```python
import math
import cmath

# The number pi
print(f"π ≈ {math.pi:.15f}")

# Square roots of non-perfect squares are irrational
print(f"√2 ≈ {math.sqrt(2):.15f}")
print(f"√3 ≈ {math.sqrt(3):.15f}")

# The imaginary unit: sqrt(-1), which we'll revisit shortly
print(f"√(-1) is the imaginary unit i (or j in Python)")
```

## 2.2 Exponents and Logarithms

Exponentials and logarithms appear everywhere in quantum mechanics — from wave amplitudes to probability decays. Understanding them intuitively is key.

### What are exponents?

An exponent tells you how many times to multiply a number by itself. For example:
- $2^3 = 2 \times 2 \times 2 = 8$ (we say "2 to the power of 3")
- $5^2 = 5 \times 5 = 25$ (we say "5 squared" — this comes from calculating the area of a square)
- $3^4 = 3 \times 3 \times 3 \times 3 = 81$

But exponents go beyond whole numbers. You might have seen:
- $4^{1/2} = \sqrt{4} = 2$ (a half-power means square root)
- $8^{1/3} = \sqrt[3]{8} = 2$ (a third-power means cube root)
- $2^{-1} = \frac{1}{2}$ (a negative power means "take the reciprocal")

These extensions all follow the same consistent rules, which is what makes mathematics so elegant.

### The exponent rules

The three most important exponent rules are:
$$a^m \cdot a^n = a^{m+n}, \quad \frac{a^m}{a^n} = a^{m-n}, \quad (a^m)^n = a^{m \cdot n}$$

> **Why do these rules work?** Consider $a^3 \cdot a^2$. This means $(a \times a \times a) \times (a \times a)$ — we're multiplying $a$ by itself a total of $3 + 2 = 5$ times. So $a^3 \cdot a^2 = a^5$. The other rules follow similarly.

### The natural exponential function

The **natural exponential function** $e^x$ (where $e \approx 2.718$, known as *Euler's number*) is the most important function in quantum mechanics. Why? Because it naturally arises from differential equations — the mathematical language of change.

> **Where does $e$ come from?** One way to understand $e$ is through compound interest. If you invest money at 100% interest compounded continuously (meaning the interest is added at every instant, not just once a year), your money grows by a factor of $e$. This continuous growth model appears everywhere in nature: population growth, radioactive decay, cooling of hot objects, and quantum wavefunctions.

```python
# Exponentials
base, exp1, exp2 = 2, 3, 5

print(f"{base}^{exp1} = {base ** exp1}")        # 2 × 2 × 2 = 8
print(f"{base}^{exp2} = {base ** exp2}")        # 2 × 2 × 2 × 2 × 2 = 32

# Verify exponent rule: a^m * a^n = a^(m+n)
# When we multiply two powers with the same base, we add the exponents
print(f"\n{base}^{exp1} × {base}^{exp2} = {base ** exp1 * base ** exp2}")  # 8 × 32 = 256
print(f"{base}^({exp1}+{exp2}) = {base ** (exp1 + exp2)}")       # 2^8 = 256 ✓

# This rule works because:
# 2^3 × 2^5 = (2×2×2) × (2×2×2×2×2) = 2^8  (we count 8 twos total)
```

### Logarithms

The natural logarithm $\ln(x)$ is the inverse of $e^x$, meaning $\ln(e^x) = x$. You can think of it as answering the question: "To what power must we raise $e$ to get $x$?"

If $e^3 \approx 20.085$, then $\ln(20.085) \approx 3$.

Logarithms are the mathematical equivalent of "undoing" an exponential. Just as subtraction undoes addition and division undoes multiplication, the logarithm undoes exponentiation.

> **Why logarithms in quantum mechanics?** Many quantum processes involve exponential growth or decay (such as radioactive decay or the probability amplitude of finding a particle in a potential well). When we want to extract information from these exponential relationships, logarithms are our primary tool.

#### Common logarithmic rules
$$\ln(ab) = \ln(a) + \ln(b) \qquad \ln\!\left(\frac{a}{b}\right) = \ln(a) - \ln(b) \qquad \ln(a^n) = n\,\ln(a)$$

These rules are incredibly powerful — they let us convert multiplication into addition, which is why logarithmic scales (like the Richter scale for earthquakes and decibels for sound) are so useful in science.

```python
# Example: finding what power of e gives us x
x = math.exp(3)   # e^3 ≈ 20.086
log_x = math.log(x)  # natural log: "to what power must we raise e to get x?"

print(f"e^3 ≈ {x:.6f}")         # e cubed is about 20.086
print(f"ln(e^3) = {log_x}")     # taking the natural log undoes the exponential: answer is 3
print(f"So ln and e^ cancel each other out: {abs(log_x - 3) < 1e-10}")  # YES!

# Log rules: ln(ab) = ln(a) + ln(b)
# This is the logarithmic equivalent of the exponent rule: a^m × a^n = a^(m+n)
a, b = 5, 7
print(f"\nln({a} × {b}) = ln({a * b}) = {math.log(a * b):.6f}")    # ln(35)
print(f"ln({a}) + ln({b}) = {math.log(a) + math.log(b):.6f}")      # ln(5) + ln(7)
# These are equal! Logarithms turn multiplication into addition.
```

## 2.3 Algebra and Equations

Algebra lets us solve for unknown quantities by manipulating equations while keeping them balanced. Think of an equation like a set of scales: whatever you do to one side, you must do to the other. In quantum mechanics, we solve equations constantly — to find the energy levels of electrons, the probabilities of different outcomes, and the behaviour of waves.

### Solving Linear Equations

The simplest equation is linear: it involves only first powers of the unknown. Given $ax + b = c$, we solve for $x$:
1. Subtract $b$ from both sides: $ax = c - b$
2. Divide by $a$: $x = \frac{c - b}{a}$

That's it! The key is to isolate $x$ by performing the *inverse* operations in the reverse order that they were applied.

```python
# Solve ax + b = c for x: x = (c - b) / a
a, b, c = 3, 7, 16

# Step-by-step:
#   3x + 7 = 16
#   3x     = 16 - 7     (subtract 7 from both sides)
#   3x     = 9
#   x      = 9 / 3      (divide both sides by 3)
#   x      = 3

x = (c - b) / a
print(f"{a}x + {b} = {c} → x = (c - b) / a = ({c} - {b}) / {a} = {x}")

# Verify by plugging x back into the original equation
print(f"Check: {a}({x}) + {b} = {a * x + b}")  # Should equal 16
```

### Quadratic Equations

Quadratic equations are slightly more challenging because they involve $x^2$. They appear everywhere — from calculating the trajectory of a ball to determining the energy levels in quantum systems.

Given $ax^2 + bx + c = 0$, the **quadratic formula** gives:
$$x = \frac{-b \pm \sqrt{b^2 - 4ac}}{2a}$$

> **Where does this formula come from?** It's derived by a technique called "completing the square" — a brilliant algebraic trick. We add and subtract a carefully chosen number to turn the left side into a perfect square. While the derivation is interesting, what matters most is understanding the *parts* of the formula:
> - The $-b$ shifts the result based on the linear term.
> - The $\pm$ means there are **two** solutions (unless they happen to be the same).
> - The square root introduces the possibility of complex numbers when the discriminant is negative.
> - The $2a$ at the bottom normalises the result.

#### The Discriminant: $\Delta = b^2 - 4ac$

The discriminant tells us what *kind* of solutions we'll get:
- $\Delta > 0$: Two distinct **real** solutions (the parabola crosses the x-axis twice).
- $\Delta = 0$: One **repeated** solution (the parabola just touches the x-axis).
- $\Delta < 0$: Two **complex** solutions (the parabola never crosses the x-axis).

The third case is particularly important in quantum mechanics, because complex numbers are the language of quantum amplitudes.

```python
# Solve ax^2 + bx + c = 0
a, b, c = 1, -5, 6  # x^2 - 5x + 6 = 0, roots at x=2 and x=3

discriminant = b**2 - 4*a*c
root1 = (-b + math.sqrt(discriminant)) / (2*a)
root2 = (-b - math.sqrt(discriminant)) / (2*a)

print(f"{a}x² + ({b})x + {c} = 0")
print(f"Discriminant: Δ = b² - 4ac = {discriminant}")
print(f"Root 1: x = {root1}")
print(f"Root 2: x = {root2}")

# Step-by-step:
#   1) Calculate the discriminant: Δ = (-5)² - 4(1)(6) = 25 - 24 = 1
#   2) Take the square root: √1 = 1
#   3) Apply the formula: x = (5 ± 1) / 2
#   4) Root 1 = (5 + 1) / 2 = 3, Root 2 = (5 - 1) / 2 = 2

# Verify by plugging roots back into the original equation
print(f"\nCheck root 1: {a}*({root1})² + {b}*({root1}) + {c} = {a*root1**2 + b*root1 + c}")
print(f"Check root 2: {a}*({root2})² + {b}*({root2}) + {c} = {a*root2**2 + b*root2 + c}")
# Both should equal 0 — confirming our answers are correct!

# Try changing a, b, c to try different quadratic equations!
# What happens when the discriminant is negative?
```

### What if the discriminant is negative?

When $b^2 - 4ac < 0$, the square root involves **negative numbers**, which leads us to **complex numbers** — the language of quantum amplitudes.

> **A quick puzzle:** What number, when multiplied by itself, gives $-1$? No real number satisfies this, because a positive times a positive is positive, and a negative times a negative is also positive. This is why mathematicians invented the **imaginary unit**, denoted $i$ (or $j$ in engineering and Python), where $i = \sqrt{-1}$.

With this single invention, we can now take the square root of *any* number. And remarkably, this seemingly abstract concept turns out to be essential for describing how quantum particles behave.

```python
# Solve x^2 + 4 = 0, which has no real solution
a2, b2, c2 = 1, 0, 4  # x^2 + 4 = 0

discriminant2 = b2**2 - 4*a2*c2
print(f"{a2}x² + {b2}x + {c2} = 0")
print(f"Discriminant: Δ = {discriminant2} (negative → complex roots)")

# Step-by-step:
#   1) Δ = 0² - 4(1)(4) = -16
#   2) √(-16) = √(16) × √(-1) = 4i
#   3) x = (0 ± 4i) / 2 = ±2i
# So the two roots are x = 2i and x = -2i

# In Python, we create complex numbers using the 'j' suffix (engineering notation)
root1 = complex(0, 2)  # 0 + 2i
root2 = complex(0, -2) # 0 - 2i
print(f"Roots: x = {root1} and x = {root2}")

# Verify: (2i)² = 4i² = 4(-1) = -4, so x² + 4 = -4 + 4 = 0 ✓
print(f"\nVerification: (2i)² = {(2j)**2}")
print(f"So (2i)² + 4 = {(2j)**2 + 4}")  # Should equal 0
```

## 2.4 Functions

A **function** $f(x)$ is like a machine: you put in a number, and it gives you back exactly one output. The input is called the *argument* (usually $x$), and the output is the *value* of the function at that input.

> **Everyday analogy:** Think of a function like a vending machine. You press button A1 (input), and exactly one bag of crisps comes out (output). You can't get two different snacks from the same button, and you can't get nothing at all. That's what "exactly one output" means.

In quantum mechanics, **wavefunctions** are functions — typically of position $\psi(x)$ or time $\psi(t)$. The wavefunction contains all the information we can know about a quantum system.

Common functions you'll encounter include:
- **Polynomials** (like $x^2$, $x^3 + 2x - 1$)
- **Exponentials** (like $e^x$, $e^{-x^2}$)
- **Trigonometric functions** (like $\sin x$, $\cos x$)
- **Absolute value** (like $|x|$)
- **Logarithms** (like $\ln x$)

```python
# Polynomial function: f(x) = x² - 2x + 1
# This can be factored as f(x) = (x - 1)² — a parabola that touches the x-axis at x = 1

def f(x):
    return x**2 - 2*x + 1

print("f(x) = x² - 2x + 1:")
print("Notice how the outputs form a parabola shape:")
for x in [-2, -1, 0, 1, 2, 3]:
    print(f"  f({x:>2}) = {f(x):>2}")

# Key observations:
# - The minimum value is at x = 1, where f(1) = 0 (this is where the parabola "turns around")
# - The function is symmetric about x = 1: f(1 + d) = f(1 - d) for any distance d
#   For example: f(0) = f(2) = 1 and f(-1) = f(3) = 4

# Try factoring: x² - 2x + 1 = (x - 1)(x - 1) = (x - 1)²
# This means the only solution to f(x) = 0 is x = 1 (a "repeated root")
```

### Key properties of functions

Understanding these properties helps us classify and work with functions in quantum mechanics:

- **Even functions:** $f(-x) = f(x)$ — the graph is *symmetric* about the y-axis. Imagine folding the graph along the y-axis; both halves match perfectly. Common examples include $f(x) = x^2$ and $f(x) = \cos x$.

- **Odd functions:** $f(-x) = -f(x)$ — the graph is *symmetric* about the origin. If you rotate the graph by 180° around the origin, it looks the same. Common examples include $f(x) = x^3$ and $f(x) = \sin x$.

- **Periodic functions:** $f(x + T) = f(x)$ for some period $T$ — the graph repeats itself every $T$ units along the x-axis. The sine and cosine functions are periodic with period $2\pi \approx 6.283$. This repeating behaviour is crucial for describing waves.

> **Why care about even and odd functions?** In quantum mechanics, many wavefunctions are either even or odd. This symmetry tells us about the physical properties of the system, such as whether the particle is equally likely to be found on the left or right side of a potential well.

```python
# Test even/odd properties
def even_test(f, x):
    return f(-x) == f(x)

def odd_test(f, x):
    return f(-x) == -f(x)

x = 3
print(f"At x = {x}:")
print(f"  Even: f(x) = x² → f(-x) = {x**2}, f(-x) == f(x)? {even_test(lambda t: t**2, x)}")
print(f"  Odd:  f(x) = x³ → f(-x) = {(-x)**3}, f(-x) == -f(x)? {odd_test(lambda t: t**3, x)}")
```

## 2.5 Trigonometry

Trigonometric functions describe oscillations and waves — the heart of quantum mechanics. Every quantum particle behaves like a wave, and sine and cosine are the mathematical tools we use to describe that wave.

> **Why trigonometry?** You might remember SOH-CAH-TOA from school: sine, cosine, and tangent relate the angles of a right-angled triangle to the ratios of its sides. But this is just the starting point. In higher mathematics, we extend trigonometry to the **unit circle**, which lets us handle *any* angle — not just the acute angles in a triangle. This extension is what makes trigonometry powerful enough for quantum mechanics.

### The Unit Circle

On the unit circle (a circle with radius 1, centred at the origin), any point at angle $\theta$ measured counter-clockwise from the positive x-axis has coordinates:
$$(\cos\theta, \sin\theta)$$

This definition generalises what you learned in school. When $\theta = 30°$ (or $\pi/6$ radians), $\cos(30°) \approx 0.866$ and $\sin(30°) = 0.5$. When $\theta = 90°$ (or $\pi/2$ radians), we get $\cos(90°) = 0$ and $\sin(90°) = 1$.

> **Key concept: Radians vs degrees.** In mathematics and physics, angles are measured in **radians**, not degrees. One radian is the angle subtended when the arc length equals the radius. Since the full circumference of a unit circle is $2\pi$, a full rotation is $2\pi$ radians = 360°. Thus, $180° = \pi$ radians, $90° = \pi/2$ radians, and so on. You'll see radians everywhere in quantum mechanics.

```python
# Angles in radians
angles = [0, math.pi/6, math.pi/4, math.pi/3, math.pi/2, math.pi]
angles_names = ["0", "30°", "45°", "60°", "90°", "180°"]

print(f"{'Angle':<10} {'Radians':<15} {'sin':<10} {'cos':<10}")
print("-" * 45)
for name, angle in zip(angles_names, angles):
    sin_val = round(math.sin(angle), 10)
    cos_val = round(math.cos(angle), 10)
    # Clean up representation
    sin_str = "0" if sin_val == 0 else f"{sin_val:.4f}"
    cos_str = "1" if angle == 0 else f"{cos_val:.4f}"
    print(f"  {name:<8}   {angle:.4f}     {sin_str:<10} {cos_str:<10}")
```

### The Pythagorean Identity

The most important trigonometric identity follows directly from the Pythagorean theorem applied to the unit circle:
$$\sin^2\theta + \cos^2\theta = 1$$

Here's why this is true: on the unit circle, any point $(\cos\theta, \sin\theta)$ is exactly distance 1 from the origin (that's what "unit circle" means!). By the Pythagorean theorem, the distance from the origin is $\sqrt{\cos^2\theta + \sin^2\theta} = 1$, so squaring both sides gives the identity.

> **Why does this matter in quantum mechanics?** This identity is the foundation for understanding normalisation — the requirement that the total probability of finding a particle *somewhere* must equal 1. The same mathematical structure appears: if you have two probabilities that add up to 1, they obey the same relationship as sine and cosine squared.

#### Other useful trigonometric identities

From this fundamental identity, we can derive others:
- $\tan\theta = \frac{\sin\theta}{\cos\theta}$
- $1 + \tan^2\theta = \sec^2\theta$
- $\sin(2\theta) = 2\sin\theta\cos\theta$
- $\cos(2\theta) = \cos^2\theta - \sin^2\theta$

These become useful when simplifying wavefunction equations.

```python
# Verify the Pythagorean identity
theta = math.pi / 7  # Any angle works
sin_sq = math.sin(theta)**2
cos_sq = math.cos(theta)**2

print(f"θ = π/7 ≈ {theta:.4f}")
print(f"sin²(θ) = {sin_sq:.10f}")
print(f"cos²(θ) = {cos_sq:.10f}")
print(f"sin²(θ) + cos²(θ) = {sin_sq + cos_sq:.10f}")
print(f"Identity holds: {abs(sin_sq + cos_sq - 1) < 1e-10}")
```

### Euler's Formula — The Bridge to Quantum Mechanics

This is arguably the most important equation in all of quantum mechanics:
$$e^{i\theta} = \cos\theta + i\sin\theta$$

It connects three seemingly unrelated areas of mathematics — **exponentials**, **trigonometry**, and **complex numbers** — into one elegant relationship. To appreciate how remarkable this is, consider that each of these topics is usually taught in a completely separate maths class. Euler's formula shows they are deeply intertwined.

#### What does this formula *mean*?

Imagine a point rotating around the unit circle in the complex plane. The angle of rotation is $\theta$. Euler's formula tells us that this rotation can be described compactly as $e^{i\theta}$ — an exponential with an imaginary exponent. As $\theta$ changes, the point moves around the circle, tracing out $\cos\theta + i\sin\theta$.

> **A visual way to think about it:** 
> - The real part of $e^{i\theta}$ (the horizontal coordinate) is $\cos\theta$
> - The imaginary part of $e^{i\theta}$ (the vertical coordinate) is $\sin\theta$
> - The distance from the origin is always 1 (it's on the unit circle)

#### Euler's Identity: The Special Case

When $\theta = \pi$, we get:
$$e^{i\pi} = \cos\pi + i\sin\pi = -1 + i(0) = -1$$

Rearranging gives the famous **Euler's identity**:
$$e^{i\pi} + 1 = 0$$

This equation is remarkable because it links together five of the most important numbers in mathematics:
- $e$ (from calculus and exponential growth)
- $i$ (from complex numbers)
- $\pi$ (from geometry and circles)
- $1$ (the multiplicative identity)
- $0$ (the additive identity)

In one elegant equation!

```python
# Euler's formula
theta = math.pi / 3

# Left side: e^(iθ)
euler_lhs = cmath.exp(1j * theta)

# Right side: cos(θ) + i·sin(θ)
euler_rhs = math.cos(theta) + 1j * math.sin(theta)

print(f"θ = π/3 = {theta:.4f}")
print(f"e^(iθ) = {euler_lhs}")
print(f"cos(θ) + i·sin(θ) = {euler_rhs}")
print(f"They match: {abs(euler_lhs - euler_rhs) < 1e-14}")
```

```python
# Verify Euler's identity: e^(iπ) + 1 = 0
eulers_identity = cmath.exp(1j * math.pi) + 1
print(f"e^(iπ) + 1 = {eulers_identity}")
print(f"Close to zero: {abs(eulers_identity) < 1e-14}")
```

## 2.6 Probability and Statistics Basics

Quantum mechanics is inherently *probabilistic*. This is not because we lack good enough measuring instruments — it is a fundamental feature of nature itself. We cannot predict with certainty what a single measurement will give us. All we can predict are the *probabilities* of different outcomes.

> **Why is this so strange?** In our everyday experience, probabilities usually reflect our ignorance. If I flip a coin and hide it under my hand, I say "there's a 50% chance it's heads" — but the coin is *already* either heads or tails; I just don't know which. In quantum mechanics, however, the particle is genuinely in a *superposition* of states until measured. The outcome is not determined until the moment of measurement. This is what Einstein famously struggled with when he said "God does not play dice."

### Discrete Probability

For a discrete set of outcomes, if outcome $i$ has probability $p_i$, then two rules must be satisfied:
- $0 \leq p_i \leq 1$ for all $i$ — probabilities are never negative and never exceed 100%.
- $\sum_i p_i = 1$ — the total probability of *something* happening must be exactly 1 (or 100%).

#### A concrete example: rolling a die

A fair six-sided die has six possible outcomes (1, 2, 3, 4, 5, 6), each with probability $\frac{1}{6}$. If we roll the die many times, we expect each number to appear roughly one sixth of the time. But on any single roll, we cannot predict the outcome.

> **Key insight:** Even though we cannot predict individual outcomes, probability theory lets us make *very* accurate predictions about averages and distributions over many trials. This is enough for quantum mechanics to make incredibly precise predictions.

#### What if the die is not fair?

Suppose we have a loaded die where the probabilities are:
- P(1) = 0.1, P(2) = 0.1, P(3) = 0.1, P(4) = 0.1, P(5) = 0.1, P(6) = 0.5

The total still adds up to 1, but now we're much more likely to roll a 6. In quantum mechanics, the probabilities are often *not* equal — and calculating them is one of our main goals.

```python
# Example: probability distribution for a six-sided die
outcomes = [1, 2, 3, 4, 5, 6]
probs = [1/6] * 6

print("Discrete uniform distribution (fair die):")
for o, p in zip(outcomes, probs):
    print(f"  P({o}) = {p:.4f}")

print(f"\nTotal probability: {sum(probs):.10f}")
print(f"Valid probability distribution: {abs(sum(probs) - 1) < 1e-10}")
```

### Expected Value and Variance

Once we have a probability distribution, we can calculate summary statistics that tell us about the *typical* behaviour of the system.

The **expected value** (or mean) of a discrete random variable is the weighted average of all possible values, where the weights are the probabilities:
$$E[X] = \sum_i x_i \cdot p_i$$

Think of it as what you'd expect to get if you repeated the experiment many times and took the average. For a fair die, every outcome is equally likely ($\frac{1}{6}$), so the expected value is simply $\frac{1+2+3+4+5+6}{6} = 3.5$.

> **Note:** The expected value doesn't have to be one of the possible outcomes! You can never roll a 3.5 on a die, but it is still the long-run average.

The **variance** measures how spread out the outcomes are from the expected value:
$$\text{Var}(X) = E[(X - \mu)^2] = \sum_i (x_i - \mu)^2 \cdot p_i$$
where $\mu = E[X]$ is the mean.

A small variance means outcomes cluster tightly around the mean; a large variance means they are more spread out. The **standard deviation** ($\sigma = \sqrt{\text{Var}(X)}$) is simply the square root of the variance, expressed in the same units as the original values.

> **Why variance matters in quantum mechanics:** The Heisenberg uncertainty principle is fundamentally about the variance (or standard deviation) of position and momentum measurements. Even if you prepare identical quantum systems in identical states, the outcomes will vary — and the uncertainty principle puts a fundamental limit on how small the product of these variances can be.

```python
# Expected value and variance for a die
prob_dist = list(zip(outcomes, probs))

# Expected value
expected_value = sum(x * p for x, p in prob_dist)
print(f"Expected value E[X] = {expected_value}")

# Variance
variance = sum((x - expected_value)**2 * p for x, p in prob_dist)
print(f"Variance Var(X) = {variance:.4f}")
print(f"Standard deviation σ = {variance**0.5:.4f}")
```

## 2.7 Complex Numbers

Now we arrive at the mathematical essence of quantum mechanics. **Complex numbers** are not just a mathematical curiosity — they are the natural language for probability amplitudes, which are at the heart of quantum theory.

A complex number has a **real part** and an **imaginary part**:
$$z = a + ib$$
where $i = \sqrt{-1}$ is the imaginary unit and $a, b \in \mathbb{R}$ are real numbers.

> **A note on terminology:** Many people think of the "imaginary" part as something made up or unreal. But complex numbers are *very* real — they describe physical phenomena with incredible precision. The name is a historical accident (introduced by Descartes as a derogatory term), and a better name would be "extended numbers" since they extend the real numbers in a perfectly sensible way.

#### Visualising complex numbers: the Complex Plane

Real numbers live on a one-dimensional number line. Complex numbers live on a **two-dimensional plane** called the *complex plane*:
- The horizontal axis represents the **real part** ($a$).
- The vertical axis represents the **imaginary part** ($b$).

So the complex number $3 + 4i$ is the point $(3, 4)$ on this plane. This visualisation is incredibly helpful for understanding operations like multiplication and conjugation.

#### Polar form: magnitude and phase

Using Euler's formula, we can write a complex number in **polar form**:
$$z = r e^{i\theta} = r(\cos\theta + i\sin\theta)$$

where:
- $r = |z| = \sqrt{a^2 + b^2}$ is the **magnitude** (the distance from the origin to the point). This is also called the *modulus*.
- $\theta$ is the **phase angle** (the angle the line from the origin makes with the positive real axis, measured in radians). This is also called the *argument*.

> **Why two forms?** The Cartesian form ($a + ib$) is convenient for addition and subtraction. The polar form ($re^{i\theta}$) is convenient for multiplication, division, and powers. In quantum mechanics, the phase is crucial — it determines how quantum states interfere with each other.

```python
# Complex number arithmetic
z1 = 3 + 4j
z2 = 1 - 2j

print(f"z1 = {z1}")
print(f"z2 = {z2}\n")

print(f"Addition (z1 + z2): {z1 + z2}")
print(f"Multiplication (z1 * z2): {z1 * z2}")

# Convert to polar form
r1 = abs(z1)
theta1 = cmath.phase(z1)
print(f"\nz1 in polar form: r = {r1:.4f}, θ = {theta1:.4f} rad")

# Verify conversion back: r·e^(iθ) = r·cos(θ) + i·r·sin(θ)
reconstructed = r1 * cmath.exp(1j * theta1)
print(f"Verification: {reconstructed} ≈ {z1}")
print(f"Match: {abs(reconstructed - z1) < 1e-14}")
```

### The Complex Conjugate

The complex conjugate of $z = a + ib$ is $z^* = a - ib$. Geometrically, it is the reflection of $z$ across the real axis (the horizontal axis) on the complex plane.

Key property: multiplying a complex number by its conjugate gives a purely real, non-negative number:
$$z \cdot z^* = (a + ib)(a - ib) = a^2 - (ib)^2 = a^2 + b^2 = |z|^2$$

This is incredibly useful because it lets us "extract" the magnitude of a complex number. Notice that this is the same as the Pythagorean theorem: the magnitude squared is the sum of the squares of the real and imaginary parts.

> **Why the conjugate matters in quantum mechanics:** When we calculate the probability of a quantum outcome, we multiply the amplitude by its complex conjugate (this is Born's rule, explained below). The conjugate ensures the result is always a real, non-negative number — because probabilities must be real and non-negative, never imaginary or negative.

```python
# Complex conjugate
print(f"z1 = {z1}")
print(f"z1* (conjugate) = {z1.conjugate()}")
print(f"z1 · z1* = {z1 * z1.conjugate()}")
print(f"|z1|² = {z1.real**2 + z1.imag**2}")
print(f"In magnitude form: |z1| = {abs(z1)}")

# The result is purely real: 3² + 4² = 25
print(f"\nCheck: 3² + 4² = 9 + 16 = {3**2 + 4**2}")
```

### Born's Rule and Probability

**Born's Rule** (Max Born, 1926) states that the probability of finding a quantum system in a specific state is the magnitude squared of its probability amplitude:
$$P = |\psi|^2 = \psi \cdot \psi^*$$

> **The big idea:** In quantum mechanics, the state of a system is described by a *complex number* called a probability amplitude (usually denoted $\psi$). You cannot observe this amplitude directly — it is not a physical quantity you can measure with an instrument. However, the *probability* of observing a particular outcome is the squared magnitude of this amplitude.

Let's break down why this is so profound:

1. **Amplitudes can interfere.** Because amplitudes are complex numbers, they can have different phases. When two amplitudes add up, their phases matter: if the phases align (both pointing in roughly the same direction), they *constructively* interfere and the probability is larger. If the phases oppose (pointing in opposite directions), they *destructively* interfere and the probability is smaller. This is the mathematical mechanism behind the double-slit interference pattern.

2. **Probabilities are always real and non-negative.** Even though amplitudes are complex, the probability $|\psi|^2 = \psi \cdot \psi^*$ is always a real, non-negative number — exactly what we need for a physical probability.

3. **The amplitude carries phase information.** Two states with the same magnitude but different phases are physically distinct. This phase difference is what makes quantum mechanics different from classical probability theory.

> **Historical note:** Max Born developed this rule while trying to understand the meaning of the wavefunction. He received the Nobel Prize in Physics in 1954 largely for this insight. Before Born, even Einstein and Schrödinger thought the wavefunction described a physical charge or mass density. Born realised it was about *probability* — a revolutionary idea at the time.

#### Normalisation condition

For a single particle, the total probability of finding it *somewhere* must be 1 (certainty). This means:
$$\sum_i |\psi_i|^2 = 1$$

This is why we say quantum states must be "normalised" — the sum of all probabilities must equal 1. It's the same mathematical structure as the Pythagorean identity: $\sin^2\theta + \cos^2\theta = 1$.

```python
# Born's rule example
amplitude = (3 + 4j) / 5  # normalized so |amplitude|² = 1

probability = amplitude * amplitude.conjugate()

print(f"Probability amplitude: {amplitude}")
print(f"Probability = |amplitude|² = {probability.real:.4f}")
print(f"Magnitude: |amplitude| = {abs(amplitude):.4f}")
print(f"\nThe amplitude is normalized: |amplitude|² = 1.0 (as expected)")
```

## 2.8 Vectors and Matrices (Linear Algebra)

By this point, we've covered arithmetic, algebra, functions, trigonometry, complex numbers, and probability. Now comes the final piece: **linear algebra** — the study of vectors and matrices. Linear algebra is the primary mathematical language of quantum mechanics, and it might feel very different from the algebra you learned in school.

> **A shift in perspective:** In school algebra, you solve for *numbers* (like $x = 5$). In linear algebra, you solve for *vectors* and *matrices*. Instead of asking "what number satisfies this equation?" you ask "what vector satisfies this equation?" This shift in thinking is essential for quantum mechanics.

A quantum state is represented as a **vector** in a complex vector space (called a **Hilbert space**). Physical operations (like measuring energy or applying a magnetic field) are represented by **matrices** that transform these vectors.

### Column Vectors and Ket Notation

Physicist Paul Dirac invented a brilliant notation called **bra-ket notation** (pronounced "bra" and "ket") to keep track of vectors and their partners:
- A **ket** $|\psi\rangle$ represents a column vector (think of it as the right half of a bracket $\rangle$).
- A **bra** $\langle\psi|$ represents the conjugate transpose — a row vector containing the complex conjugates (think of it as the left half of a bracket $\langle$).

Together, $\langle\phi|\psi\rangle$ forms a "bracket" representing the inner product (like a dot product) between two vectors.

The two fundamental qubit states — the quantum analogue of 0 and 1 in classical computing — are written as:

```python
import numpy as np

# Define the computational basis states
ket_0 = np.array([[1],
                  [0]])
ket_1 = np.array([[0],
                  [1]])

print("State |0> = ", ket_0.T)
print("State |1> = ", ket_1.T)
print("\nWritten as column vectors:")
print("|0⟩ = [1, 0]^T")
print("|1⟩ = [0, 1]^T")
```

### Inner Products (Bra·Ket)

The inner product $\langle\phi|\psi\rangle$ tells us how "aligned" two states are — it's the quantum version of asking "how similar are these two vectors?"

> **Analogy:** Imagine two arrows on a piece of paper. If they point in exactly the same direction, they are perfectly aligned. If they point in perpendicular directions, they are completely unrelated. The inner product quantifies this idea mathematically.

Key properties:
- **Orthogonal states:** If $\langle\phi|\psi\rangle = 0$, the states are *orthogonal* (perpendicular). In quantum mechanics, this means the states are completely distinguishable. If the system is in state $|\psi\rangle$, a measurement will *never* find it in state $|\phi\rangle$.
- **Normalised states:** If $\langle\psi|\psi\rangle = 1$, the state is normalised — the total probability of finding the particle *somewhere* is 100%.
- **Orthonormal sets:** If $\langle i|j\rangle = \delta_{ij}$ (the Kronecker delta, equal to 1 when $i=j$ and 0 otherwise), the states form an orthonormal basis — the quantum analogue of the x and y axes in 2D space.

For the qubit states $|0\rangle$ and $|1\rangle$:
- $\langle 0|0\rangle = 1$ (|0⟩ is normalised to itself)
- $\langle 1|1\rangle = 1$ (|1⟩ is normalised to itself)
- $\langle 0|1\rangle = 0$ (|0⟩ and |1⟩ are orthogonal — they are perfectly distinguishable)

This orthonormality is fundamental: it guarantees that when we measure a qubit, we get exactly one definite outcome (either 0 or 1), never both and never neither.

```python
# The bra is the conjugate transpose of the ket
# In numpy: H is conjugate transpose (Hermitian transpose)
bra_0 = ket_0.T.conj()  # <0|
bra_1 = ket_1.T.conj()  # <1|

# Inner products
overlap_01 = bra_0 @ ket_1  # <0|1>
overlap_00 = bra_0 @ ket_0  # <0|0>
overlap_11 = bra_1 @ ket_1  # <1|1>

print(f"<0|0> = {overlap_00} (orthonormal: yes)")
print(f"<1|1> = {overlap_11} (orthonormal: yes)")
print(f"<0|1> = {overlap_01} (orthogonal: yes)")
```

### Vector Norm (Normalization)

The **norm** (pronounced "norm" as in "standard") of a vector measures its length — how far it stretches from the origin. For a quantum state, the norm must equal 1. This is not just a mathematical convenience; it has a deep physical meaning:

> **The normalisation condition:** $\langle\psi|\psi\rangle = 1$ means the total probability of finding the particle *somewhere* is 100%. If the norm were less than 1, there would be less than 100% chance of finding the particle. If it were greater than 1, the probabilities would be nonsense.

For a vector $|\psi\rangle = \begin{pmatrix} a \\ b \end{pmatrix}$, the norm squared is:
$$\langle\psi|\psi\rangle = |a|^2 + |b|^2$$

This is simply the sum of the probabilities of all possible outcomes — and it must equal 1.

#### How to normalise a vector

If you have a vector that isn't normalised, you can fix it by dividing by its norm. For example, the vector $\begin{pmatrix} 3 \\ 4 \end{pmatrix}$ has norm $\sqrt{3^2 + 4^2} = \sqrt{25} = 5$. Dividing by 5 gives the normalised vector $\begin{pmatrix} 3/5 \\ 4/5 \end{pmatrix} = \begin{pmatrix} 0.6 \\ 0.8 \end{pmatrix}$.

> **Connection to the Pythagorean theorem:** The normalisation condition $|a|^2 + |b|^2 = 1$ is the same structure as the Pythagorean identity $\sin^2\theta + \cos^2\theta = 1$. If we write $a = \cos\theta$ and $b = \sin\theta$, the normalisation condition is automatically satisfied. This is no coincidence — both express the idea that "the parts add up to the whole".

```python
# Check normalization of |0⟩ and |1⟩
norm_0 = np.sqrt(np.real((bra_0 @ ket_0).real))[0, 0][0, 0]
norm_1 = np.sqrt(np.real((bra_1 @ ket_1).real))[0, 0][0, 0]
print(f"||0⟩|| = {norm_0:.4f}")
print(f"||1⟩|| = {norm_1:.4f}")
print(f"Both normalized: {abs(norm_0 - 1) < 1e-10 and abs(norm_1 - 1) < 1e-10}")
```

### Superposition

**Superposition** is perhaps the most famous (and most misunderstood) concept in quantum mechanics. It means that a quantum system can exist in a combination of multiple states simultaneously — not as one *or* the other, but as *both at the same time*.

A quantum system in superposition is written as:
$$|\psi\rangle = \alpha |0\rangle + \beta |1\rangle$$

where $\alpha$ and $\beta$ are complex numbers called **probability amplitudes**, and they must satisfy the normalisation condition:
$$|\alpha|^2 + |\beta|^2 = 1$$

> **The cat analogy (and why it's misleading):** Erwin Schrödinger proposed a thought experiment involving a cat in a sealed box with a radioactive atom. If the atom decays, the cat dies; if it doesn't, the cat lives. According to quantum mechanics, until we open the box, the cat is in a *superposition* of alive and dead. While this is a popular illustration, it's actually a *reductio ad absurdum* — Schrödinger meant it to show how absurd it sounds, not to claim cats can actually be both alive and dead. Quantum superpositions are very fragile and don't scale up to everyday objects like cats.

#### What does "both at once" really mean?

When we say the qubit is in a superposition of $|0\rangle$ and $|1\rangle$, we mean:
1. The state *exists* as a definite mathematical object: $|\psi\rangle = \alpha |0\rangle + \beta |1\rangle$.
2. When we *measure* the qubit, we only ever see one outcome: either 0 (with probability $|\alpha|^2$) or 1 (with probability $|\beta|^2$).
3. The superposition is revealed when we perform *multiple* measurements on identically prepared systems — the statistics match the probabilities $|\alpha|^2$ and $|\beta|^2$.

> **Key insight:** The superposition is real, but the act of measurement *collapses* it to one definite outcome. Before measurement, the qubit is genuinely in both states. After measurement, it settles into whichever outcome was observed. This collapse is one of the deepest mysteries of quantum mechanics.

#### The role of phase in superposition

Notice that $\alpha$ and $\beta$ are *complex* numbers — they have both a magnitude and a phase. The phase doesn't affect the probabilities of individual outcomes (since $|\alpha|^2$ only depends on the magnitude), but it *does* affect how different parts of the superposition *interfere* with each other when operations are applied. This interference is what gives quantum computers their power.

```python
# Create a superposition state
alpha = 1 / np.sqrt(2)
beta = 1j / np.sqrt(2)  # beta has a phase (imaginary part)

# Verify normalization: |α|² + |β|² = 1
norm_sq = abs(alpha)**2 + abs(beta)**2
print(f"Superposition: |ψ⟩ = {alpha}|0⟩ + {beta}i|1⟩")
print(f"Normality check: |α|² + |β|² = {norm_sq:.4f}")

# The ket vector
ket_psi = alpha * ket_0 + beta * ket_1
print(f"\n|ψ> = {ket_psi.flatten()}")

# Probabilities of each outcome
p0 = abs(alpha)**2
p1 = abs(beta)**2
print(f"\nIf we measure:")
print(f"  P(0) = |α|² = {p0:.4f} → {p0 * 100:.0f}%")
print(f"  P(1) = |β|² = {p1:.4f} → {p1 * 100:.0f}%")
print(f"  Total = {p0 + p1:.4f}")
```

### Quantum Operators (Matrices)

In quantum mechanics, **operations** — such as measuring energy, applying a magnetic field, or flipping a qubit — are represented by **matrices**. These matrices transform one quantum state into another, just like a function transforms one number into another.

> **A shift in perspective:** In school, you probably encountered functions like $f(x) = x^2$. In quantum mechanics, the "functions" are *linear operators* — matrices that act on vectors. When we apply an operator to a state vector, we get a new state vector. This is the mathematical language of "doing something" to a quantum system.

#### What does a matrix actually *do*?

A matrix is just a rectangular array of numbers. But when you multiply a matrix by a vector, it transforms the vector — it can stretch it, rotate it, reflect it, or squish it. Here are some intuitive examples:

- **Scaling:** The identity matrix $\begin{pmatrix} 1 & 0 \\ 0 & 1 \end{pmatrix}$ does nothing — it leaves every vector unchanged. It's like multiplying by 1.
- **Swapping:** The Pauli-X matrix $\begin{pmatrix} 0 & 1 \\ 1 & 0 \end{pmatrix}$ flips the vector — it swaps the top and bottom components. If the input is $|0\rangle$, the output is $|1\rangle$, and vice versa. This is the quantum equivalent of a "NOT" gate.
- **Rotation:** Some matrices rotate vectors in the complex plane. These are crucial for creating and manipulating superpositions.

#### Eigenvalues and eigenvectors

When a matrix operator $M$ acts on a special kind of vector (called an **eigenvector**), it doesn't change the direction — it only stretches or shrinks the vector by a factor called the **eigenvalue**:
$$M|\lambda\rangle = \lambda |\lambda\rangle$$

> **Why this matters in quantum mechanics:** The eigenvectors of an operator represent the states in which a measurement gives a *definite* result (no superposition). The eigenvalues are the possible measurement outcomes. For example, the energy levels of an atom are the eigenvalues of the Hamiltonian (energy) operator.

#### Physical observables as Hermitian operators

In quantum mechanics, every measurable quantity (called an **observable**) — energy, position, momentum, spin — is represented by a **Hermitian matrix**. Hermitian matrices have two crucial properties:
1. All eigenvalues are **real numbers** — measurement outcomes must be real, not complex.
2. Eigenvectors corresponding to different eigenvalues are **orthogonal** — different outcomes are perfectly distinguishable.

The Pauli matrices (X, Y, Z) are the most important Hermitian operators in quantum computing, and they represent the three spatial directions in which we can measure a qubit's spin.

```python
# Pauli-X matrix
X_gate = np.array([[0, 1],
                   [1, 0]])

# Apply X to |0>
X_on_0 = X_gate @ ket_0
print("X|0⟩ =", X_on_0.flatten())
print("Expected: |1⟩ = [0, 1]")
print(f"Match: {np.array_equal(np.round(X_on_0.flatten(), 10), ket_1.flatten())}")

# Apply X to |1>
X_on_1 = X_gate @ ket_1
print(f"\nX|1⟩ = {X_on_1.flatten()}")
print("Expected: |0⟩ = [1, 0]")
print(f"Match: {np.array_equal(np.round(X_on_1.flatten(), 10), ket_0.flatten())}")
```

### The Hadamard Gate

The **Hadamard gate** is perhaps the most important quantum gate because it is the standard way to *create* a superposition from a definite state. It is denoted by the letter $H$ and is defined as:
$$H = \frac{1}{\sqrt{2}} \begin{pmatrix} 1 & 1 \\ 1 & -1 \end{pmatrix}$$

> **Why the factor of $\frac{1}{\sqrt{2}}$?** This normalisation factor ensures that the output states are properly normalised (have unit length). Without it, the probabilities wouldn't add up to 1, which would violate the fundamental rules of quantum mechanics.

When applied to $|0\rangle$, the Hadamard gate creates an *equal* superposition:
$$H|0\rangle = \frac{|0\rangle + |1\rangle}{\sqrt{2}}$$

This means if we measure the qubit, we have a 50% chance of getting 0 and a 50% chance of getting 1. It's the quantum analogue of flipping a fair coin.

When applied to $|1\rangle$, it creates a *different* equal superposition:
$$H|1\rangle = \frac{|0\rangle - |1\rangle}{\sqrt{2}}$$

Notice the **minus sign** in front of $|1\rangle$. This is not just a cosmetic difference — the minus sign represents a *phase difference*. In quantum mechanics, phase matters enormously. Two superpositions that look similar can have very different behaviours because of subtle phase differences. This is the source of quantum interference, which is exploited in quantum algorithms to amplify correct answers and cancel wrong ones.

#### The Hadamard transform is its own inverse

Applying the Hadamard gate twice brings you back to where you started:
$$H \cdot H = I$$
where $I$ is the identity matrix. This is a special property — not all quantum gates are reversible in this way. It also means that if you apply $H$ to a superposition state, you can "undo" the superposition and recover the original definite state.

> **Why is this important?** It means quantum operations are *reversible* — you can always go backwards. This is fundamentally different from classical computing, where operations like AND and OR are irreversible (you can't tell what the inputs were just by looking at the output). Reversibility is a key feature of quantum mechanics and has deep implications for quantum computing and thermodynamics.

```python
# Hadamard gate
H = (1 / np.sqrt(2)) * np.array([[1, 1],
                                 [1, -1]])

H_on_0 = H @ ket_0
H_on_1 = H @ ket_1

print("H|0⟩ =", H_on_0.flatten())
print("H|1⟩ =", H_on_1.flatten())

# Verify both are normalized
norm_H0 = np.sqrt(np.real(H_on_0.T.conj() @ H_on_0))
norm_H1 = np.sqrt(np.real(H_on_1.T.conj() @ H_on_1))
print(f"\n||H|0⟩|| = {norm_H0:.4f}")
print(f"||H|1⟩|| = {norm_H1:.4f}")
```
